```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

def run_backtest(strategy_weights_dict, clean_i, parquet_path, start_date, end_date, ic_logs_dict=None, fee_rate=0.001, save_path='.'):
    print(f"🚀 启动多策略对比回测: {start_date} 至 {end_date}...")
    
    df_stock = pd.read_parquet(parquet_path)
    df_stock['TradingDate'] = pd.to_datetime(df_stock['TradingDate']).dt.normalize()
    df_stock['Symbol'] = df_stock['Symbol'].astype(str).str.zfill(6)
    
    start, end = pd.to_datetime(start_date), pd.to_datetime(end_date)
    all_trading_days = sorted(df_stock['TradingDate'].unique())
    all_trading_days = [d for d in all_trading_days if start <= d <= end]
    
    # --- 构造等权重策略 ---
    df_temp = df_stock[df_stock['TradingDate'].isin(all_trading_days)][['TradingDate', 'Symbol']]
    ew_counts = df_temp.groupby('TradingDate')['Symbol'].transform('count')
    df_temp['weight'] = 1.0 / ew_counts
    df_temp.columns = ['date', 'symbol', 'weight']
    
    strategy_weights_dict = {**{'Equal_Weight': df_temp}, **strategy_weights_dict}

    all_strategies_ret = pd.DataFrame(index=all_trading_days)
    all_strategies_turnover = {}

    for name, clean_w in strategy_weights_dict.items():
        print(f"正在处理策略: {name}...")
        clean_w = clean_w.copy()
        clean_w['date'] = pd.to_datetime(clean_w['date']).dt.normalize()
        clean_w['symbol'] = clean_w['symbol'].astype(str).str.zfill(6)
        
        rebalance_dates = sorted(clean_w['date'].unique())
        rebalance_map = {d: clean_w[clean_w['date'] == d].set_index('symbol')['weight'].to_dict() for d in rebalance_dates}

        current_weights = pd.Series(dtype=float)
        strat_daily_ret = pd.Series(index=all_trading_days, dtype=float)
        total_turnover = 0.0
        
        init_candidates = [d for d in rebalance_dates if d <= all_trading_days[0]]
        if init_candidates:
            init_date = max(init_candidates)
            current_weights = pd.Series(rebalance_map[init_date])
            current_weights = current_weights / current_weights.sum()
            total_turnover += 1.0

        for i in range(len(all_trading_days) - 1):
            t, t_next = all_trading_days[i], all_trading_days[i+1]
            turnover_cost = 0 

            if t in rebalance_map:
                new_weights_target = pd.Series(rebalance_map[t])
                new_weights_target = new_weights_target / new_weights_target.sum()
                w_old, w_new = current_weights.align(new_weights_target, fill_value=0)
                turnover = (w_new - w_old).abs().sum() 
                total_turnover += turnover
                turnover_cost = turnover * fee_rate 
                current_weights = w_new 

            if current_weights.empty: continue
            day_stock_data = df_stock[df_stock['TradingDate'] == t_next].set_index('Symbol')
            arithmetic_rets = day_stock_data['Simple_Ret'].reindex(current_weights.index).fillna(0)
            strat_daily_ret[t_next] = (current_weights * arithmetic_rets).sum() - turnover_cost
            current_weights = current_weights * (1 + arithmetic_rets)
            if current_weights.sum() != 0: current_weights /= current_weights.sum()
        
        all_strategies_ret[name] = strat_daily_ret.fillna(0)
        actual_years = (pd.to_datetime(end) - pd.to_datetime(start)).days / 365.25
        all_strategies_turnover[name] = total_turnover / actual_years if actual_years > 0 else 0

    clean_i['date'] = pd.to_datetime(clean_i['date']).dt.normalize()
    curr_index = clean_i.set_index('date')['index_ret'].reindex(all_trading_days).fillna(0)
    all_strategies_ret['Benchmark'] = curr_index
    all_strategies_turnover['Benchmark'] = 0

    nv_df = (1 + all_strategies_ret).cumprod()
    _plot_and_report_multi(nv_df, all_strategies_ret, all_strategies_turnover, ic_logs_dict, save_path=save_path)

def _plot_and_report_multi(nv, res, turnover_dict, ic_logs_dict=None, save_path='.'):
    if not os.path.exists(save_path):
        os.makedirs(save_path)

    # --- 1. 累计净值与回撤图 ---
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10), sharex=True, gridspec_kw={'height_ratios': [2, 1]})
    for col in nv.columns:
        lw, ls = (2.0, '-') if col not in ['Benchmark', 'Equal_Weight'] else (1.5, '--')
        ax1.plot(nv[col], label=col, lw=lw, ls=ls)
    ax1.set_title('Performance Comparison (Cumulative Net Value)', fontsize=15)
    ax1.legend(); ax1.grid(True, alpha=0.3)
    
    main_strat = nv.columns[0] 
    dd = (nv[main_strat] / nv[main_strat].cummax()) - 1
    ax2.fill_between(dd.index, dd, 0, alpha=0.2, color='red', label=f'MDD {main_strat}')
    ax2.set_ylabel('Drawdown'); ax2.grid(True, alpha=0.3)
    ax2.legend()
    plt.tight_layout()
    plt.savefig(os.path.join(save_path, 'backtest_cumulative_performance.png'), dpi=300)
    plt.show()

    # --- 新增: 累计超额收益 (Alpha) 曲线图 ---
    plt.figure(figsize=(14, 6))
    for col in nv.columns:
        if col != 'Benchmark':
            # 累计超额收益 = 策略净值 / 基准净值
            excess_nv = nv[col] / nv['Benchmark']
            plt.plot(excess_nv, label=f'{col} vs Benchmark', lw=2)
    plt.axhline(1.0, color='red', linestyle='--', alpha=0.6)
    plt.title('Cumulative Excess Return (Strategy Net Value / Benchmark Net Value)', fontsize=15)
    plt.ylabel('Alpha Ratio')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig(os.path.join(save_path, 'backtest_cumulative_excess_return.png'), dpi=300)
    plt.show()

    # --- 2. 分年度收益统计与绘图准备 ---
    yearly_stats = (1 + res).resample('YE').prod() - 1
    yearly_stats.index = yearly_stats.index.year
    
    x = np.arange(len(yearly_stats))
    width = 0.8 / len(nv.columns) # 自动计算柱状图宽度
    
    # --- 3. 绘图: 年度绝对收益对比 ---
    plt.figure(figsize=(14, 6))
    for i, col in enumerate(nv.columns):
        plt.bar(x + (i - len(nv.columns)/2) * width, yearly_stats[col] * 100, width=width, label=col)
    
    plt.axhline(0, color='black', linewidth=1)
    plt.ylabel('Annual Return (%)')
    plt.title('Yearly Absolute Returns Comparison')
    plt.xticks(x, yearly_stats.index)
    plt.legend()
    plt.grid(axis='y', alpha=0.3)
    plt.tight_layout()
    plt.savefig(os.path.join(save_path, 'backtest_yearly_absolute_returns.png'), dpi=300)
    plt.show()

    # --- 4. 绘图: 年度超额收益 (Alpha) 对比 ---
    plt.figure(figsize=(14, 6))
    alpha_cols = [c for c in nv.columns if c != 'Benchmark']
    width_alpha = 0.8 / len(alpha_cols)
    for i, col in enumerate(alpha_cols):
        alpha_val = (yearly_stats[col] - yearly_stats['Benchmark']) * 100
        plt.bar(x + (i - len(alpha_cols)/2) * width_alpha, alpha_val, width=width_alpha, label=f'{col} Alpha')
    
    plt.axhline(0, color='red', linestyle='--', linewidth=1)
    plt.ylabel('Excess Return vs Benchmark (%)')
    plt.title('Yearly Alpha (Active Return) Comparison')
    plt.xticks(x, yearly_stats.index)
    plt.legend()
    plt.grid(axis='y', alpha=0.3)
    plt.tight_layout()
    plt.savefig(os.path.join(save_path, 'backtest_yearly_alpha_returns.png'), dpi=300)
    plt.show()

    # --- 5. 打印各项指标表格 ---
    actual_years = (nv.index.max() - nv.index.min()).days / 365.25
    metrics = ['Final Net Value', 'Annual Return', 'Annual Volatility', 'Sharpe Ratio', 'Information Ratio', 'Max Drawdown', 'Annual Turnover']
    if ic_logs_dict: metrics += ['Mean IC', 'ICIR']

    bench_final_nv = nv['Benchmark'].iloc[-1]
    bench_cagr = (bench_final_nv)**(1/actual_years) - 1

    all_stats = {}
    for col in nv.columns:
        risk_free_rate = 0.02
        final_nv = nv[col].iloc[-1]
        cagr = (final_nv)**(1/actual_years) - 1
        vol = res[col].std() * np.sqrt(242)
        sharpe = (cagr - risk_free_rate) / vol if vol != 0 else 0
        mdd = ((nv[col] / nv[col].cummax()) - 1).min()
        
        if col == 'Benchmark':
            ir_str = "-"
        else:
            active_ret_daily = res[col] - res['Benchmark']
            tracking_error = active_ret_daily.std() * np.sqrt(242)
            ir = (cagr - bench_cagr) / tracking_error if tracking_error != 0 else 0
            ir_str = f"{ir:.2f}"
        
        stats = [f"{final_nv:.4f}", f"{cagr:.2%}", f"{vol:.2%}", f"{sharpe:.2f}", ir_str, f"{mdd:.2%}", f"{turnover_dict.get(col, 0):.2f}x"]
        
        if ic_logs_dict and col in ic_logs_dict:
            ic_series = ic_logs_dict[col].iloc[:, 2] 
            m_ic = ic_series.mean()
            s_ic = ic_series.std()
            icir = m_ic / s_ic if s_ic != 0 else 0
            stats += [f"{m_ic:.4f}", f"{icir:.4f}"]
        elif ic_logs_dict:
            stats += ["-", "-"]
        all_stats[col] = stats

    header = f"{'Performance Metric':<20}"
    for col in nv.columns: header += f" | {col:<12}"
    line = "=" * len(header)
    print(f"\n{line}\n{header}\n{'-' * len(header)}")
    for i, m in enumerate(metrics):
        row = f"{m:<20}"
        for col in nv.columns: row += f" | {all_stats[col][i]:>12}"
        print(row)
    print(line)

    # --- 6. 打印年度收益数据表格 ---
    print("\n📅 Yearly Annualized Return Breakdown")
    y_header = f"{'Year':<20}"
    for col in nv.columns: y_header += f" | {col:<12}"
    print("-" * len(y_header))
    print(y_header)
    print("-" * len(y_header))
    for year, row_data in yearly_stats.iterrows():
        row_str = f"{year:<20}"
        for col in nv.columns: row_str += f" | {row_data[col]:>12.2%}"
        print(row_str)
    print("-" * len(y_header))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

# --- 1. 设置路径 ---
# 确保 6 个模型的路径均正确
paths = {
    'Ridge': r"../weights_diagnostic/old/岭回归3/Ridge_MVO_DynamicPM_Weights.xlsx",
    'XGBoost': r"../weights_diagnostic/old/XGBoost3/XGB_DynamicPM_Weights.xlsx",
    'LightGBM': r"../weights_diagnostic/old/LightGBM3/LGBM_DynamicPM_Weights.xlsx",
    'LSTM_VSN': r"../weights_diagnostic/old/LSTM_VSN/LSTM_VSN_MVO_Weights.xlsx",
    'Ensemble_Eqw': r"../weights_diagnostic/old/集成模型3/HS300_EW_Ensemble_Final.xlsx",
    'Ensemble_CEnet': r"../weights_diagnostic/old/集成模型3/HS300_CENet_Dynamic_Final.xlsx"
}

# 读取行业映射数据
# 建议确保 hs300_daily_data.parquet 包含最新的行业分类
df_industry = pd.read_parquet('hs300_daily_data.parquet')
df_industry['TradingDate'] = pd.to_datetime(df_industry['TradingDate']).dt.normalize()
df_industry['Symbol'] = df_industry['Symbol'].astype(str).str.zfill(6)

def plot_six_models_with_industry(target_date, save_path='.'):
    # 检查并创建保存路径
    if not os.path.exists(save_path):
        os.makedirs(save_path)

    target_date = pd.to_datetime(target_date).normalize()
    
    # 修改为 2行 3列，并调整画布大小
    fig, axes = plt.subplots(2, 3, figsize=(24, 14))
    axes = axes.flatten()
    
    # 提取目标日期的行业快照 (假设列名为 Indexcode 或 Industryname)
    industry_snap = df_industry[df_industry['TradingDate'] == target_date].set_index('Symbol')['Indexcode'].to_dict()
    
    for i, (name, path) in enumerate(paths.items()):
        if not os.path.exists(path):
            axes[i].text(0.5, 0.5, f'File Not Found:\n{name}', ha='center', va='center', color='red')
            continue

        # A. 读取权重
        df = pd.read_excel(path)
        df['date'] = pd.to_datetime(df['date']).dt.normalize()
        df['symbol'] = df['symbol'].astype(str).str.zfill(6)
        
        # B. 提取目标日持仓
        day_data = df[df['date'] == target_date].sort_values('weight', ascending=False)
        
        if day_data.empty:
            axes[i].text(0.5, 0.5, f'No data for {name}\non {target_date.date()}', ha='center', va='center')
            axes[i].set_title(f'{name} Allocation', fontsize=14, fontweight='bold')
            continue
            
        # C. 匹配行业
        day_data['industry'] = day_data['symbol'].map(industry_snap).fillna('Unknown')
        
        # D. 整理前 10 大持仓
        top_n = 10
        plot_df = day_data.head(top_n).copy()
        
        # 构造标签: "代码 (行业)"
        plot_df['label'] = plot_df.apply(lambda r: f"{r['symbol']}\n({r['industry']})", axis=1)
        
        others_weight = day_data.iloc[top_n:]['weight'].sum()
        if others_weight > 0:
            others_row = pd.DataFrame({'label': ['Others'], 'weight': [others_weight]})
            plot_df = pd.concat([plot_df, others_row], ignore_index=True)
        
        # E. 绘图 (环形图)
        wedges, texts, autotexts = axes[i].pie(
            plot_df['weight'], 
            labels=plot_df['label'], 
            autopct='%1.1f%%', 
            startangle=140,
            pctdistance=0.75, # 调整百分比位置，防止重叠
            colors=plt.cm.tab20c(np.linspace(0, 1, len(plot_df))), # 使用颜色更丰富的 Tab20
            wedgeprops={'width': 0.45, 'edgecolor': 'w'} 
        )
        
        axes[i].set_title(f'{name}', fontsize=18, fontweight='bold', pad=10)
        plt.setp(autotexts, size=9, weight="bold")
        plt.setp(texts, size=9) 

    # 整体标题
    plt.suptitle(f'Cross-Model Portfolio Allocation Analysis: {target_date.date()}', 
                 fontsize=26, y=1.02, fontweight='bold')
    
    plt.tight_layout()
    
    # 保存图片
    save_file = os.path.join(save_path, f'full_6_model_allocation_{target_date.date()}.png')
    plt.savefig(save_file, dpi=300, bbox_inches='tight')
    print(f"🚀 六模型持仓对比图已保存至: {save_file}")
    
    plt.show()

# --- 执行 ---
# 选择 2024-12-31 做最终截面分析
target_day = '2024-09-26' 
plot_six_models_with_industry(target_day)
```

```python
import pandas as pd
import numpy as np

# --- 1. 设置路径 ---
weight_paths = {
    'Ridge': r"../weights_diagnostic/old/岭回归3/Ridge_MVO_DynamicPM_Weights.xlsx",
    'XGBoost': r"../weights_diagnostic/old/XGBoost3/XGB_DynamicPM_Weights.xlsx",
    'LightGBM': r"../weights_diagnostic/old/LightGBM3/LGBM_DynamicPM_Weights.xlsx",
    'LSTM_VSN': r"../weights_diagnostic/old/LSTM_VSN/LSTM_VSN_MVO_Weights.xlsx",
    'Ensemble_Eqw': r"../weights_diagnostic/old/集成模型3/HS300_EW_Ensemble_Final.xlsx",
    'Ensemble_CENet': r"../weights_diagnostic/old/集成模型3/HS300_CENet_Dynamic_Final.xlsx"
}
clean_index_file = r"../RAW_DATA/沪深300指数收益率/standard_index_2024.xlsx"

# 目标日期设为 2024-09-27
target_date = pd.to_datetime('2024-09-26').normalize()

# --- 2. 计算各模型收益 (从对数转换为百分比) ---
# 提取行情 df 中的 Log_Ret
returns_snap = df[pd.to_datetime(df['TradingDate']).dt.normalize() == target_date].set_index('Symbol')['Log_Ret']
results = []

for name, path in weight_paths.items():
    w_df = pd.read_excel(path)
    w_df['date'] = pd.to_datetime(w_df['date']).dt.normalize()
    w_df['symbol'] = w_df['symbol'].astype(str).str.zfill(6)
    
    day_weights = w_df[w_df['date'] == target_date].set_index('symbol')['weight']
    
    if not day_weights.empty:
        # 对齐权重与当日 Log_Ret
        combined = pd.concat([day_weights, returns_snap], axis=1, join='inner')
        
        # 计算组合的对数收益率
        port_log_return = (combined['weight'] * combined['Log_Ret']).sum()
        
        # 转换为百分比收益率 (Arithmetic Return)
        port_pct_return = (np.exp(port_log_return) - 1)
        
        results.append({
            'Model': name,
            'Daily_Return_Pct': port_pct_return * 100  # 以百分号数值形式存储
        })

# --- 3. 读取指数收益率 (无需转化，直接使用) ---
index_df = pd.read_excel(clean_index_file)
index_df['date'] = pd.to_datetime(index_df['date']).dt.normalize()

# 获取当日指数收益
target_index_data = index_df[index_df['date'] == target_date]

if not target_index_data.empty:
    target_index_ret = target_index_data['index_ret'].values[0]
    # 假设 index_ret 本身就是 0.04 表示 4%，如果是 4 表示 4%，请根据数据微调
    # 这里统一存为百分数 (例如 4.5 表示 4.5%)
    results.append({
        'Model': 'HS300_Benchmark',
        'Daily_Return_Pct': target_index_ret * 100 
    })
else:
    print(f"⚠️ Index data not found for {target_date.date()}")

# --- 4. 汇总展示 ---
performance_df = pd.DataFrame(results).sort_values('Daily_Return_Pct', ascending=False)

print(f"\n--- Performance Comparison on {target_date.date()} (Unit: %) ---")
print(performance_df.to_string(index=False))

# 计算超额百分比收益 (Alpha in %)
benchmark_val = performance_df[performance_df['Model'] == 'HS300_Benchmark']['Daily_Return_Pct'].values[0]
performance_df['Alpha_Pct'] = performance_df['Daily_Return_Pct'] - benchmark_val

print("\n--- Alpha (Excess Return over Benchmark in %) ---")
print(performance_df[['Model', 'Alpha_Pct']].sort_values('Alpha_Pct', ascending=False))
```
