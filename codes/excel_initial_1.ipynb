```python
import pandas as pd

# 请确保路径前带 r，防止 unicode 转义错误
input_path = r"../CSI300_CHANGE_WIND/2010-2024沪深300成分股.xlsx"
output_path = r"../CSI300_CHANGE_WIND/2010-2024沪深300成分股_处理后.xlsx"

def process_wind_excel(input_file, output_file):
    excel_file = pd.ExcelFile(input_file)
    all_sheets = excel_file.sheet_names
    
    with pd.ExcelWriter(output_file, engine='openpyxl') as writer:
        for sheet_name in all_sheets:
            # 关键点：header=5 表示以第 6 行作为列名
            # 如果第 6 行还不对，可以尝试 header=4
            df = pd.read_excel(input_file, sheet_name=sheet_name, header=5)
            
            # 清理列名中的空格（Wind 导出有时带空格）
            df.columns = [str(c).strip() for c in df.columns]
            
            # 寻找包含 "Wind代码" 的列
            target_col = None
            possible_names = ['Wind代码', 'wind_code', '代码']
            for name in possible_names:
                for col in df.columns:
                    if name in col:
                        target_col = col
                        break
                if target_col: break
            
            if target_col:
                # 提取前 6 位数字代码
                # 处理可能存在的空值，确保转换为字符串
                df['代码(无后缀)'] = df[target_col].astype(str).str.extract(r'(\d{6})')
                
                # 将处理后的结果保存
                df.to_excel(writer, sheet_name=sheet_name, index=False)
                print(f"✅ 已成功处理 Sheet: {sheet_name}")
            else:
                # 打印调试信息，看看现在的列名到底是什么
                print(f"❌ 跳过 Sheet: {sheet_name}。当前识别到的列名为: {list(df.columns[:3])}")
                df.to_excel(writer, sheet_name=sheet_name, index=False)

process_wind_excel(input_path, output_path)
print("\n--- 全部处理完成 ---")
```

```python
import pandas as pd
import os

# 1. 设置路径
input_path = r"../CSI300_CHANGE_WIND/2010-2024沪深300成分股_处理后.xlsx"
output_folder = r"../CSI300_CHANGE_WIND/成分股按日期拆分_修正版"

if not os.path.exists(output_folder):
    os.makedirs(output_folder)

def split_excel_with_zeros(input_file, folder):
    excel_file = pd.ExcelFile(input_file)
    all_sheets = excel_file.sheet_names
    
    for sheet_name in all_sheets:
        # 关键点 1：dtype=str 强制所有数据按字符串读取，不让 Python 把 000001 变成 1
        df = pd.read_excel(input_file, sheet_name=sheet_name, dtype=str)
        
        # 这里的列名请根据你上一步生成的实际列名对齐，假设是 '代码(无后缀)'
        target_col = '代码(无后缀)' 
        
        if target_col in df.columns:
            # 提取列
            new_df = df[[target_col]].copy()
            
            # 关键点 2：万一读取时还是丢了 0，用 zfill(6) 补齐到 6 位
            new_df[target_col] = new_df[target_col].astype(str).str.zfill(6)
            
            # 重命名为 code
            new_df.columns = ['code']
            
            # 保存
            file_name = f"{sheet_name}.xlsx"
            save_path = os.path.join(folder, file_name)
            
            # 关键点 3：虽然 index=False，但保存后的 Excel 打开时，
            # Excel 软件本身可能还会显示为数字，这不代表数据错了。
            new_df.to_excel(save_path, index=False)
            print(f"已生成并补零: {file_name}")
        else:
            print(f"⚠️ 跳过 {sheet_name}")

split_excel_with_zeros(input_path, output_folder)
print("\n--- 处理完成，前导0已强制保留 ---")
```
