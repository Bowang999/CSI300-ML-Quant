```python
import pandas as pd
import numpy as np
import os

# ==========================================
# 1. 加载原始行情数据 (300只股票/日)
# ==========================================
print("正在读取原始行情数据...")
df_stock = pd.read_parquet('stock_data.parquet')

# 确保日期和代码格式统一
df_stock['TradingDate'] = pd.to_datetime(df_stock['TradingDate']).dt.normalize()
df_stock['Symbol'] = df_stock['Symbol'].astype(str).str.zfill(6)

# 清理可能存在的旧指标（如果之前运行过，防止列名冲突）
cols_to_drop = [c for c in ['PE', 'PB', 'PCF', 'PS'] if c in df_stock.columns]
if cols_to_drop:
    df_stock = df_stock.drop(columns=cols_to_drop)

# ==========================================
# 2. 读取并整合 8 个衍生指标 Excel
# ==========================================
# 假设你的 8 个文件放在 'derivative_data' 文件夹下
# 如果在当前目录，可以修改路径
folder_path = r'../RAW_DATA/股票衍生指标' 
all_files = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.endswith('.xlsx')]

df_deriv_list = []

print(f"正在读取 {len(all_files)} 个衍生指标文件...")
for file in all_files:
    # 读取时跳过 CSMAR 的中文表头行（第二、三行）
    temp_df = pd.read_excel(file)
    
    # 清洗：强制转换日期，剔除文字行
    temp_df['TradingDate'] = pd.to_datetime(temp_df['TradingDate'], errors='coerce', format='mixed')
    temp_df = temp_df.dropna(subset=['TradingDate'])
    
    # 格式化：代码补全 6 位
    temp_df['Symbol'] = temp_df['Symbol'].astype(str).str.zfill(6)
    
    # 只取需要的列，防止 ShortName 等重复
    keep_cols = ['TradingDate', 'Symbol', 'PE', 'PB', 'PCF', 'PS']
    # 动态检查文件中包含哪些指标（防止 8 个文件指标分布不均）
    actual_cols = [c for c in keep_cols if c in temp_df.columns]
    
    df_deriv_list.append(temp_df[actual_cols])

# 合并所有 Excel 数据
df_deriv_all = pd.concat(df_deriv_list, axis=0).drop_duplicates(subset=['TradingDate', 'Symbol'])

# ==========================================
# 3. 执行精确对齐合并 (Left Join)
# ==========================================
print("执行多对一精确合并...")
# 以 df_stock 为准，新表中的 780 只股票中不属于 300 只的部分会被自动过滤
df_final = pd.merge(
    df_stock, 
    df_deriv_all, 
    on=['TradingDate', 'Symbol'], 
    how='left'
)

# ==========================================
# 4. 保存与检查
# ==========================================
print(f"合并后总行数: {len(df_final)}")
deriv_cols = ['PE', 'PB', 'PCF', 'PS']
print(f"衍生指标缺失统计：\n{df_final[deriv_cols].isna().sum()}")

df_final.to_parquet('stock_data_with_derivatives.parquet', index=False)
print("--- 衍生指标合并完成！ ---")
```

```python
# 1. 加载合并后的数据
df = pd.read_parquet('stock_data_with_derivatives.parquet')
df.tail(10)
```

```python
import pandas as pd

# 2. 定义需要检查的衍生指标列
# 如果你的列名在 Excel 里是中文，请确保这里对应修改（例如：'市盈率'）
deriv_cols = ['PE', 'PB', 'PCF', 'PS'] 

# 3. 整体缺失率统计
print("--- 整体缺失情况统计 ---")
total_rows = len(df)
missing_summary = pd.DataFrame({
    '缺失行数': df[deriv_cols].isna().sum(),
    '缺失比例 (%)': (df[deriv_cols].isna().mean() * 100).round(2)
})
print(missing_summary)

# 4. 分年度缺失率统计（核心：检查历史数据断层）
# 确保 TradingDate 是日期格式
df['Year'] = pd.to_datetime(df['TradingDate']).dt.year

print("\n--- 分年度指标缺失率 (%) ---")
# 计算每年每个指标的 NaN 比例
yearly_missing = df.groupby('Year')[deriv_cols].apply(lambda x: (x.isna().mean() * 100).round(2))
print(yearly_missing)

# 5. 异常诊断：找出完全缺失指标的股票
print("\n--- 严重缺失预警 (缺失率 > 50% 的股票) ---")
stock_missing = df.groupby('Symbol')[deriv_cols].apply(lambda x: x.isna().mean())
serious_missing = stock_missing[(stock_missing > 0.5).any(axis=1)]

if not serious_missing.empty:
    print(f"检测到 {len(serious_missing)} 只股票存在严重数据缺失：")
    print(serious_missing.head(10)) # 仅展示前 10 只
else:
    print("恭喜！所有成分股的衍生指标覆盖度良好。")

# 6. (可选) 查看特定年份的样本数量，确认 300 只股票是否都在
print("\n--- 每年观测值数量 (期望值: 300 * 交易日数) ---")
print(df.groupby('Year')['Symbol'].count())
```
