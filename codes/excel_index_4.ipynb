```python
import pandas as pd

# 1. 路径设置
input_parquet = r"stock_data.parquet"
output_excel = r"../CSI300_CHANGE_WIND/all_stock_symbols_list.xlsx"

def export_all_symbols():
    print("正在从行情数据中提取所有股票代码...")
    # 读取行情主表
    df = pd.read_parquet(input_parquet)
    
    # 2. 提取所有唯一的股票代码
    all_symbols = df['Symbol'].unique()
    
    # 3. 构建 DataFrame
    # 按照你的要求：列名为 code
    symbols_df = pd.DataFrame({'code': all_symbols})
    
    # 4. 排序（从小到大，方便核对）
    symbols_df = symbols_df.sort_values('code')
    
    # 5. 导出
    symbols_df.to_excel(output_excel, index=False)
    
    print("-" * 30)
    print(f"✅ 导出成功！")
    print(f"你的数据集中共有 {len(symbols_df)} 只不同的股票。")
    print(f"清单已保存至: {output_excel}")
    print("建议：在 CSMAR 下载时，时间范围请设为 1990-01-01 到 2026-12-31。")

if __name__ == "__main__":
    export_all_symbols()
```

```python
import pandas as pd
import numpy as np

# ==========================================
# 1. 加载数据
# ==========================================
print("正在读取数据...")
df_stock = pd.read_parquet(r"stock_data_backward.parquet")
df_industry = pd.read_excel(r"../RAW_DATA/行业分类/总行业/总行业.xlsx")

# 2. 清洗行业表
# 仅保留中证行业分类 ID (P0206, P0215, P0219)
csi_ids = ['P0206', 'P0215', 'P0219']
df_csi = df_industry[df_industry['IndustryClassificationID'].isin(csi_ids)].copy()

# 强制转换日期
df_stock['TradingDate'] = pd.to_datetime(df_stock['TradingDate'], errors='coerce')
df_csi['ImplementDate'] = pd.to_datetime(df_csi['ImplementDate'], errors='coerce')

# 强制转换 Symbol 为 6 位字符串
df_stock['Symbol'] = df_stock['Symbol'].astype(str).str.zfill(6)
df_csi['Symbol'] = df_csi['Symbol'].astype(str).str.zfill(6)

# 剔除无效行
df_stock = df_stock.dropna(subset=['TradingDate', 'Symbol']).copy()
df_csi = df_csi.dropna(subset=['ImplementDate', 'Symbol', 'IndustryCode']).copy()

# 3. 核心逻辑：使用索引映射（避开 merge_asof 的排序陷阱）
print("正在构建行业变动字典...")
# 将行业信息处理成：{Symbol: [(日期1, 行业1), (日期2, 行业2)]}
industry_map = {}
for name, group in df_csi.sort_values('ImplementDate').groupby('Symbol'):
    industry_map[name] = group[['ImplementDate', 'IndustryCode']].values.tolist()

def get_industry(row):
    symbol = row['Symbol']
    date = row['TradingDate']
    
    if symbol not in industry_map:
        return None
    
    # 获取该股票的所有行业变更记录
    changes = industry_map[symbol]
    
    # 寻找小于等于当前日期的最新行业
    last_code = None
    for change_date, code in changes:
        if date >= change_date:
            last_code = code
        else:
            break
            
    # 如果此时 last_code 还是 None，说明当前日期早于第一条记录日期
    # 为了满足你的需求（2015前覆盖），我们取该股票的第一条记录
    if last_code is None:
        last_code = changes[0][1]
        
    return last_code

# 4. 执行映射（100万行大约需要 20-40 秒）
print("正在为行情匹配行业 (这可能需要一点时间)...")
df_stock['Indexcode'] = df_stock.apply(get_industry, axis=1)

# 5. 最后检查：对于完全没匹配到的（新股等），进行全局填充
print("正在执行最终填充...")
df_stock['Indexcode'] = df_stock.groupby('Symbol')['Indexcode'].ffill().bfill()

# 6. 保存
output_name = 'stock_with_industry_safe.parquet'
df_stock.to_parquet(output_name, index=False)

print(f"处理完成！成功生成: {output_name}")
print(df_stock.head())
```

```python
import pandas as pd

# 1. 加载你处理好的最终文件
df = pd.read_parquet('stock_with_industry_safe.parquet')

# 2. 提取两只股票的数据
vanke = df[df['Symbol'] == '000002'][['TradingDate', 'Indexcode']].rename(columns={'Indexcode': 'Vanke_Industry'})
baoan = df[df['Symbol'] == '000009'][['TradingDate', 'Indexcode']].rename(columns={'Indexcode': 'BaoAn_Industry'})

# 3. 合并在一起对比
comparison = pd.merge(vanke, baoan, on='TradingDate', how='inner')

# 4. 找出它们行业代码不一致的日期
diff = comparison[comparison['Vanke_Industry'] != comparison['BaoAn_Industry']]

if diff.empty:
    print("在整个时间段内，这两只股票的行业代码始终保持一致。")
else:
    first_diff_date = diff['TradingDate'].min()
    print(f"首次出现行业分化的日期是: {first_diff_date.date()}")
    print("-" * 50)
    print("分化前后的对比示例：")
    
    # 打印分化前最后 3 天和分化后前 5 天
    pre_diff = comparison[comparison['TradingDate'] < first_diff_date].tail(3)
    post_diff = comparison[comparison['TradingDate'] >= first_diff_date].head(5)
    
    print(pd.concat([pre_diff, post_diff]))

# 5. 统计各自变动的频率
print("-" * 50)
print("万科 A 经历的行业代码变动：")
print(vanke.drop_duplicates(subset=['Vanke_Industry']))

print("\n中国宝安 经历的行业代码变动：")
print(baoan.drop_duplicates(subset=['BaoAn_Industry']))
```
