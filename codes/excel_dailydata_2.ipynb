```python
import pandas as pd
import glob
import os

# 1. 设置文件所在路径
# 假设你的 36 个文件都在这个目录下
input_folder = r"../RAW_DATA/全股票日行情 - 后复权" 
output_parquet = r"stock_data_backward.parquet"

def combine_stock_data(folder_path, output_file):
    file_list = glob.glob(os.path.join(folder_path, "股票日行情*.xlsx"))
    file_list.sort()
    
    all_df = []
    
    for file in file_list:
        print(f"正在处理: {os.path.basename(file)}")
        # 读取数据
        df = pd.read_excel(file, header=0)
        
        # --- 核心修复逻辑开始 ---
        # 1. 过滤掉包含中文单位的行（比如 '元/股', '股', '元' 等）
        # 我们假设 TradingDate 列如果不是日期格式，或者包含“交易日期”字样，就是干扰行
        df = df[~df['TradingDate'].astype(str).str.contains('交易日期|元/股|元|股', na=False)]
        
        # 2. 确保 Symbol 列是 6 位字符串（补零）
        df['Symbol'] = df['Symbol'].astype(str).str.extract(r'(\d+)')[0].str.zfill(6)
        
        # 3. 强制转换数值列。errors='coerce' 会把无法转换的脏数据变成 NaN
        numeric_cols = ['OpenPrice', 'ClosePrice', 'HighPrice', 'LowPrice', 'Volume', 'Amount', 'AvgPrice', 'ChangeRatio']
        for col in numeric_cols:
            if col in df.columns:
                df[col] = pd.to_numeric(df[col], errors='coerce')
        # --- 核心修复逻辑结束 ---
        
        all_df.append(df)

    print("正在合并并清洗...")
    combined_df = pd.concat(all_df, ignore_index=True)

    # 转换日期格式
    combined_df['TradingDate'] = pd.to_datetime(combined_df['TradingDate'], errors='coerce')
    
    # 剔除转换后仍为 NaT (空日期) 的行
    combined_df = combined_df.dropna(subset=['TradingDate', 'Symbol'])

    # 排序
    combined_df = combined_df.sort_values(by=['TradingDate', 'Symbol']).reset_index(drop=True)

    # 调整列顺序
    cols = combined_df.columns.tolist()
    new_cols = ['TradingDate', 'Symbol'] + [c for c in cols if c not in ['TradingDate', 'Symbol']]
    combined_df = combined_df[new_cols]

    # 保存
    combined_df.to_parquet(output_file, engine='pyarrow', index=False)
    print(f"✅ 处理完成！最终行数: {len(combined_df)}")

combine_stock_data(input_folder, output_parquet)
```

```python
df = pd.read_parquet(output_parquet)
df.head(20)
```

```python
# 1. 检查日期范围
print(f"数据起始日期: {df['TradingDate'].min()}")
print(f"数据结束日期: {df['TradingDate'].max()}")

# 2. 验证每天是否真的是 300 只股票左右
daily_counts = df.groupby('TradingDate')['Symbol'].count()
print("\n每日股票数量统计 (前10天):")
print(daily_counts.head(10))

# 3. 检查是否有缺失值
print("\n关键字段缺失值统计:")
print(df[['TradingDate', 'Symbol', 'ClosePrice']].isnull().sum())
```

```python
import pandas as pd

# 检查是否有完全相同的 日期+代码 组合
duplicates = df.duplicated(subset=['TradingDate', 'Symbol']).sum()

if duplicates > 0:
    print(f"⚠️ 注意：数据中存在 {duplicates} 条重复项！需要去重。")
else:
    print("✅ 恭喜：数据非常干净，没有任何重复的‘日期-股票’组合。")
```
