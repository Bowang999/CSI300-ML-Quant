```python
import pandas as pd
import numpy as np
# ==========================================
# 1. 数据加载
# ==========================================
print("正在读取数据...")
df_stock = pd.read_parquet('stock_with_industry_safe.parquet')
df_funda = pd.read_excel(r"../RAW_DATA/基本面数据/基本面数据1-盈利能力.xlsx")

# ==========================================
# 2. 预处理：彻底清洗 CSMAR 脏数据
# ==========================================
# A. 强制转换 Accper 为日期，无法转换的（如“没有单位”）会变成 NaT
df_funda['Accper'] = pd.to_datetime(df_funda['Accper'], errors='coerce')

# B. 剔除 Accper 为空的行（这会自动删掉表头中的文字行和单位行）
df_funda = df_funda.dropna(subset=['Accper']).copy()

# C. 筛选合并报表 (Typrep == 'A') 
# 注意：如果 Typrep 列也有中文，我们只保留值为 'A' 的行
df_funda = df_funda[df_funda['Typrep'].astype(str).str.contains('A', na=False)].copy()

# ==========================================
# 3. 构造“安全公告日”规避未来函数
# ==========================================
def get_safe_publish_date(dt):
    # 此时 dt 已经是 datetime 对象了，无需再次 pd.to_datetime
    year, month = dt.year, dt.month
    
    if month == 3:    # 一季报 -> 5月1日可用
        return pd.Timestamp(year=year, month=5, day=1)
    elif month == 6:  # 半年报 -> 9月1日可用
        return pd.Timestamp(year=year, month=9, day=1)
    elif month == 9:  # 三季报 -> 11月1日可用
        return pd.Timestamp(year=year, month=11, day=1)
    elif month == 12: # 年报 -> 次年5月1日可用
        return pd.Timestamp(year=year + 1, month=5, day=1)
    else:
        return dt + pd.DateOffset(months=4)

print("正在构造模拟公告日...")
df_funda['MatchDate'] = df_funda['Accper'].apply(get_safe_publish_date)

# ==========================================
# 4. 格式对齐与拼接
# ==========================================
# 股票代码补全
df_stock['Symbol'] = df_stock['Symbol'].astype(str).str.zfill(6)
df_funda['Stkcd'] = df_funda['Stkcd'].astype(str).str.zfill(6)
df_stock['TradingDate'] = pd.to_datetime(df_stock['TradingDate']).dt.normalize()

# 指标转数值
funda_cols = ['F050204C', 'F050504C', 'F053301C', 'F051401C', 'F051501C']
for col in funda_cols:
    df_funda[col] = pd.to_numeric(df_funda[col], errors='coerce')

# 排序
df_stock = df_stock.sort_values('TradingDate')
df_funda = df_funda.sort_values('MatchDate')

# 去重
df_funda = df_funda.drop_duplicates(subset=['Stkcd', 'MatchDate'], keep='last')

print("执行 merge_asof 拼接...")
df_final = pd.merge_asof(
    df_stock, 
    df_funda[['Stkcd', 'MatchDate', 'Accper'] + funda_cols],
    left_on='TradingDate',
    right_on='MatchDate',
    left_by='Symbol',
    right_by='Stkcd',
    direction='backward' 
)

# 保存
df_final.drop(columns=['Stkcd', 'MatchDate']).to_parquet('stock_final_cleaned.parquet', index=False)
print("--- 处理完成！---")
```

```python
# ==========================================
# 1. 加载最终数据集
# ==========================================
file_path = 'stock_data.parquet'
df = pd.read_parquet(file_path)

# ==========================================
# 2. 查看所有列名（确认结构）
# ==========================================
print("--- 数据集当前所有列名 ---")
print(df.columns.tolist())
print("-" * 30)
```

```python
# ==========================================
# 3. 基本面指标缺失值深度分析
# ==========================================
# 定义你添加的所有基本面列名
# 盈利类: F050204C(ROA), F050504C(ROE), F053301C(毛利), F051401C(利润率), F051501C(净利率)
# 现金流类: F060101C, F060201C, F060301C, F060401C
funda_cols = [
    'F050204C', 
    'F050504C', 
    'F053301C', 
    'F051401C', 
    'F051501C', 
    'F040304C', 
    'F040604C', 
    'F041405C', 
    'F041705C', 
    'F080701B', 
    'F081001B', 
    'F081601B', 
    'F060101C', 
    'F060201C', 
    'F060301C', 
    'F060401C'
]

# 过滤出确实存在的列（防止某些列名写错）
existing_funda_cols = [c for c in funda_cols if c in df.columns]

print(f"检测到基本面指标共 {len(existing_funda_cols)} 个。")

# 总体缺失统计
missing_stats = pd.DataFrame({
    '缺失行数': df[existing_funda_cols].isna().sum(),
    '缺失比例 (%)': (df[existing_funda_cols].isna().mean() * 100).round(2)
})
print("\n--- 整体缺失值统计 ---")
print(missing_stats)

# 按年份查看缺失率（识别历史断档）
df['Year'] = pd.to_datetime(df['TradingDate']).dt.year
yearly_missing = df.groupby('Year')[existing_funda_cols].apply(lambda x: x.isna().mean() * 100).round(2)
print("\n--- 分年度指标缺失率 (%) ---")
print(yearly_missing)
```

```python
# 假设 df 是你的 stock_data.parquet
df['Year'] = pd.to_datetime(df['TradingDate']).dt.year

# 选取一个核心指标，如 ROE (F050504C)
indicator = 'F050504C'

# 计算每只股票每年指标变化的次数
def count_changes(x):
    return (x.shift(1) != x).sum()

change_counts = df.groupby(['Symbol', 'Year'])[indicator].apply(count_changes)

print("--- 每年指标跳变次数分布 ---")
print(change_counts.value_counts()) 
# 正常结果：3 或 4 应该占绝大多数（对应四季报）
```

```python
import matplotlib.pyplot as plt

def plot_verification(df, symbol='000002'):
    sample = df[df['Symbol'] == symbol].sort_values('TradingDate')
    
    fig, ax1 = plt.subplots(figsize=(12, 6))

    # 绘制股价 (折线)
    ax1.plot(sample['TradingDate'], sample['ClosePrice'], color='gray', alpha=0.3, label='Price')
    ax1.set_ylabel('Stock Price')

    # 绘制财务指标 (阶梯线)
    ax2 = ax1.twinx()
    ax2.step(sample['TradingDate'], sample['F050504C'], where='post', color='blue', label='ROE (TTM)')
    ax2.set_ylabel('ROE (TTM)')

    plt.title(f'Alignment Check for {symbol}: Price vs Fundamental')
    plt.show()

plot_verification(df)
```

```python
# 计算相邻两日指标的差值
diff = df.sort_values(['Symbol', 'TradingDate']).groupby('Symbol')['F050504C'].diff()

# 统计差值不为 0 的比例
change_days_ratio = (diff != 0).mean()
print(f"指标变动天数占比: {change_days_ratio:.2%}")
# 正常结果：应该在 1.5% 左右 (即一年 250 天里只有约 4 天在变)
```

```python
# 重新运行校验，看看次数是否降下来了
df['F050504C_rounded'] = df['F050504C'].round(6)

def count_changes_v2(group):
    # 增加一个小阈值判断，只有变化超过 0.00001 才算跳变
    return (group.diff().abs() > 1e-7).sum()

new_counts = df.groupby(['Symbol', 'Year'])['F050504C_rounded'].apply(count_changes_v2)
print(new_counts.value_counts().head(10))
```
