```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from tqdm import tqdm

def run_backtest_monthly(strategy_weights_dict, clean_i, parquet_path, start_date, end_date, ic_logs_dict=None, fee_rate=0.001, save_path='.'):
    """
    月频多策略对比回测系统 - 完整图像+指标版
    包含: 累计净值图、Alpha超额曲线、年度绝对收益图、年度Alpha图、完整指标表
    """
    print(f"🚀 启动月频回测引擎: {start_date} 至 {end_date}...")
    
    # --- 1. 数据加载与预处理 ---
    if not os.path.exists(save_path):
        os.makedirs(save_path)
        
    df_stock = pd.read_parquet(parquet_path)
    df_stock['TradingDate'] = pd.to_datetime(df_stock['TradingDate']).dt.normalize()
    df_stock['Symbol'] = df_stock['Symbol'].astype(str).str.zfill(6)
    
    start, end = pd.to_datetime(start_date), pd.to_datetime(end_date)
    all_trading_days = sorted(df_stock['TradingDate'].unique())
    all_trading_days = [d for d in all_trading_days if start <= d <= end]
    
    # --- 2. 构造月频等权重策略 (Equal Weight) ---
    print("🔧 构造月频等权基准 (月末调仓模式)...")
    df_temp_all = df_stock[df_stock['TradingDate'].isin(all_trading_days)][['TradingDate', 'Symbol']]
    df_temp_all['YearMonth'] = df_temp_all['TradingDate'].dt.to_period('M')
    month_end_dates = df_temp_all.groupby('YearMonth')['TradingDate'].max().unique()
    
    df_ew_monthly = df_temp_all[df_temp_all['TradingDate'].isin(month_end_dates)].copy()
    ew_counts = df_ew_monthly.groupby('TradingDate')['Symbol'].transform('count')
    df_ew_monthly['weight'] = 1.0 / ew_counts
    df_ew_monthly = df_ew_monthly[['TradingDate', 'Symbol', 'weight']]
    df_ew_monthly.columns = ['date', 'symbol', 'weight']
    
    # 合并策略字典
    final_strategy_dict = {**{'Equal_Weight': df_ew_monthly}, **strategy_weights_dict}

    # --- 3. 核心回测循环 ---
    all_strategies_ret = pd.DataFrame(index=all_trading_days)
    all_strategies_turnover = {}

    for name, clean_w in final_strategy_dict.items():
        print(f"   正在运行策略: {name}...")
        clean_w = clean_w.copy()
        clean_w['date'] = pd.to_datetime(clean_w['date']).dt.normalize()
        clean_w['symbol'] = clean_w['symbol'].astype(str).str.zfill(6)
        
        rebalance_dates = sorted(clean_w['date'].unique())
        rebalance_map = {d: clean_w[clean_w['date'] == d].set_index('symbol')['weight'].to_dict() for d in rebalance_dates}

        current_weights = pd.Series(dtype=float)
        strat_daily_ret = pd.Series(index=all_trading_days, dtype=float)
        total_turnover = 0.0
        
        # 初始持仓
        init_candidates = [d for d in rebalance_dates if d <= all_trading_days[0]]
        if init_candidates:
            init_date = max(init_candidates)
            current_weights = pd.Series(rebalance_map[init_date])
            current_weights = current_weights / (current_weights.sum() + 1e-10)
            total_turnover += 1.0

        for i in range(len(all_trading_days) - 1):
            t, t_next = all_trading_days[i], all_trading_days[i+1]
            turnover_cost = 0 

            # 月末调仓判定
            if t in rebalance_map:
                new_weights_target = pd.Series(rebalance_map[t])
                new_weights_target = new_weights_target / (new_weights_target.sum() + 1e-10)
                w_old, w_new = current_weights.align(new_weights_target, fill_value=0)
                turnover = (w_new - w_old).abs().sum() 
                total_turnover += turnover
                turnover_cost = turnover * fee_rate 
                current_weights = w_new 

            if current_weights.empty: continue
            
            day_stock_data = df_stock[df_stock['TradingDate'] == t_next].set_index('Symbol')
            arithmetic_rets = day_stock_data['Simple_Ret'].reindex(current_weights.index).fillna(0)
            strat_daily_ret[t_next] = (current_weights * arithmetic_rets).sum() - turnover_cost
            
            # Drift 漂移
            current_weights = current_weights * (1 + arithmetic_rets)
            if current_weights.sum() != 0: current_weights /= current_weights.sum()
        
        all_strategies_ret[name] = strat_daily_ret.fillna(0)
        actual_years = (end - start).days / 365.25
        all_strategies_turnover[name] = total_turnover / actual_years if actual_years > 0 else 0

    # 基准指数处理
    clean_i['date'] = pd.to_datetime(clean_i['date']).dt.normalize()
    curr_index = clean_i.set_index('date')['index_ret'].reindex(all_trading_days).fillna(0)
    all_strategies_ret['Benchmark'] = curr_index
    all_strategies_turnover['Benchmark'] = 0

    nv_df = (1 + all_strategies_ret).cumprod()
    
    # --- 4. 统计与绘图 ---
    _plot_and_report_multi_full(nv_df, all_strategies_ret, all_strategies_turnover, ic_logs_dict, save_path)

def _plot_and_report_multi_full(nv, res, turnover_dict, ic_logs_dict=None, save_path='.'):
    # --- 图 1: 累计净值与最大回撤 ---
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10), sharex=True, gridspec_kw={'height_ratios': [2, 1]})
    for col in nv.columns:
        lw, ls = (2.5, '-') if col not in ['Benchmark', 'Equal_Weight'] else (1.5, '--')
        ax1.plot(nv[col], label=col, lw=lw, ls=ls)
    ax1.set_title('Performance Comparison (Cumulative Net Value)', fontsize=15)
    ax1.legend(); ax1.grid(True, alpha=0.3)
    
    # 以第一个非Benchmark策略画回撤
    main_strat = [c for c in nv.columns if c != 'Benchmark'][4]
    dd = (nv[main_strat] / nv[main_strat].cummax()) - 1
    ax2.fill_between(dd.index, dd, 0, alpha=0.2, color='red', label=f'MDD {main_strat}')
    ax2.set_ylabel('Drawdown'); ax2.grid(True, alpha=0.3)
    ax2.legend()
    plt.tight_layout()
    plt.savefig(os.path.join(save_path, '1_cumulative_performance.png'), dpi=300)
    plt.show()

    # --- 图 2: 累计超额收益 (Alpha曲线) ---
    plt.figure(figsize=(14, 6))
    for col in nv.columns:
        if col != 'Benchmark':
            excess_nv = nv[col] / nv['Benchmark']
            plt.plot(excess_nv, label=f'{col} vs Benchmark', lw=2)
    plt.axhline(1.0, color='red', linestyle='--', alpha=0.6)
    plt.title('Cumulative Excess Return (Alpha Ratio)', fontsize=15)
    plt.ylabel('Alpha Ratio'); plt.legend(); plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig(os.path.join(save_path, '2_cumulative_excess_return.png'), dpi=300)
    plt.show()

    # --- 年度统计准备 ---
    yearly_stats = (1 + res).resample('YE').prod() - 1
    yearly_stats.index = yearly_stats.index.year
    x = np.arange(len(yearly_stats))
    width = 0.8 / len(nv.columns)

    # --- 图 3: 年度绝对收益对比 ---
    plt.figure(figsize=(14, 6))
    for i, col in enumerate(nv.columns):
        plt.bar(x + (i - len(nv.columns)/2) * width, yearly_stats[col] * 100, width=width, label=col)
    plt.axhline(0, color='black', linewidth=1)
    plt.ylabel('Annual Return (%)'); plt.title('Yearly Absolute Returns'); plt.xticks(x, yearly_stats.index)
    plt.legend(); plt.grid(axis='y', alpha=0.3); plt.tight_layout()
    plt.savefig(os.path.join(save_path, '3_yearly_absolute_returns.png'), dpi=300)
    plt.show()

    # --- 图 4: 年度超额收益 (Alpha) ---
    plt.figure(figsize=(14, 6))
    alpha_cols = [c for c in nv.columns if c != 'Benchmark']
    width_alpha = 0.8 / len(alpha_cols)
    for i, col in enumerate(alpha_cols):
        alpha_val = (yearly_stats[col] - yearly_stats['Benchmark']) * 100
        plt.bar(x + (i - len(alpha_cols)/2) * width_alpha, alpha_val, width=width_alpha, label=col)
    plt.axhline(0, color='red', ls='--', lw=1)
    plt.ylabel('Excess Return (%)'); plt.title('Yearly Alpha Comparison'); plt.xticks(x, yearly_stats.index)
    plt.legend(); plt.grid(axis='y', alpha=0.3); plt.tight_layout()
    plt.savefig(os.path.join(save_path, '4_yearly_alpha_comparison.png'), dpi=300)
    plt.show()

    # --- 表格输出 ---
    actual_years = (nv.index.max() - nv.index.min()).days / 365.25
    metrics = ['Final Net Value', 'Annual Return', 'Annual Volatility', 'Sharpe Ratio', 'Information Ratio', 'Max Drawdown', 'Annual Turnover']
    if ic_logs_dict: metrics += ['Mean IC', 'ICIR']

    bench_cagr = (nv['Benchmark'].iloc[-1])**(1/actual_years) - 1
    all_stats = {}
    for col in nv.columns:
        cagr = (nv[col].iloc[-1])**(1/actual_years) - 1
        vol = res[col].std() * np.sqrt(242)
        sharpe = (cagr - 0.02) / vol if vol != 0 else 0
        mdd = ((nv[col] / nv[col].cummax()) - 1).min()
        ir = (cagr - bench_cagr) / ( (res[col]-res['Benchmark']).std()*np.sqrt(242) ) if col != 'Benchmark' else 0
        
        stats_list = [f"{nv[col].iloc[-1]:.4f}", f"{cagr:.2%}", f"{vol:.2%}", f"{sharpe:.2f}", 
                      f"{ir:.2f}" if col != 'Benchmark' else "-", f"{mdd:.2%}", f"{turnover_dict.get(col, 0):.2f}x"]
        
        if ic_logs_dict and col in ic_logs_dict:
            ic_series = ic_logs_dict[col].iloc[:, 1]
            stats_list += [f"{ic_series.mean():.4f}", f"{(ic_series.mean()/ic_series.std())*np.sqrt(12):.4f}"]
        elif ic_logs_dict: stats_list += ["-", "-"]
        all_stats[col] = stats_list

    # 打印绩效表
    h = f"{'Metric':<20}"
    for col in nv.columns: h += f" | {col:<12}"
    print("\n" + "="*len(h) + f"\n{h}\n" + "-"*len(h))
    for i, m in enumerate(metrics):
        row = f"{m:<20}"
        for col in nv.columns: row += f" | {all_stats[col][i]:>12}"
        print(row)
    print("="*len(h))

    # 打印年度分解表
    print("\n📅 Yearly annualized Return Breakdown")
    yh = f"{'Year':<20}"
    for col in nv.columns: yh += f" | {col:<12}"
    print("-" * len(yh) + f"\n{yh}\n" + "-" * len(yh))
    for year, row_data in yearly_stats.iterrows():
        r = f"{year:<20}"
        for col in nv.columns: r += f" | {row_data[col]:>12.2%}"
        print(r)
    print("-" * len(yh))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# 设置绘图风格，使其更符合论文质感
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['font.sans-serif'] = ['Arial']
plt.rcParams['axes.unicode_minus'] = False

def plot_monthly_model_weights(log_path, save_path='.'):
    """
    绘制月频 C-ENet 动态集成权重演化图
    针对月频数据的离散特性进行了平滑和显示优化
    """
    # 1. 数据读取与预处理
    if not os.path.exists(log_path):
        print(f"❌ 找不到日志文件: {log_path}")
        return

    df = pd.read_excel(log_path)
    df['Date'] = pd.to_datetime(df['Date'])
    df = df.sort_values('Date')
    
    # 确保权重列名与月频诊断表输出一致 (假设列名为 W_RID, W_LGB, W_XGB)
    weight_cols = ['W_RID', 'W_LGB', 'W_XGB']
    for col in weight_cols:
        if col not in df.columns:
            print(f"⚠️ 警告: 列 {col} 不在 Excel 中，请检查诊断表的列名。")
            return

    # 检查并创建保存路径
    if not os.path.exists(save_path):
        os.makedirs(save_path)
    
    # 2. 绘图设置
    plt.figure(figsize=(14, 7))
    
    # 颜色配置 (深蓝色系搭配，适合论文)
    colors = ['#2c3e50', '#c0392b', '#2980b9'] 
    labels = ['Ridge (Linear)', 'LightGBM (Non-linear)', 'XGBoost (Non-linear)']
    
    # 3. 绘制堆叠面积图 (Stacked Area Chart)
    # 月频绘图的关键：使用 'step' 绘图或在面积图上增加散点，以体现“月度离散调仓”的特征
    plt.stackplot(df['Date'], 
                  df['W_RID'], 
                  df['W_LGB'], 
                  df['W_XGB'], 
                  labels=labels,
                  alpha=0.85, 
                  colors=colors,
                  edgecolor='white', # 增加白边使层级更清晰
                  linewidth=0.5)
    
    # 4. 细节优化
    plt.title('Monthly Dynamic Model Allocation (C-ENet Weights Evolution)', fontsize=16, fontweight='bold', pad=20)
    plt.xlabel('Rebalancing Month', fontsize=12)
    plt.ylabel('Allocation Weight', fontsize=12)
    
    # 设置纵轴范围和刻度
    plt.ylim(0, 1)
    plt.yticks(np.arange(0, 1.1, 0.1))
    
    # 调整横轴日期显示，使其不拥挤
    plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%Y-%m'))
    plt.xticks(rotation=45)
    
    # 图例置于上方，避免遮挡面积
    plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), 
               ncol=3, frameon=True, shadow=True, fontsize=11)
    
    plt.grid(axis='y', linestyle='--', alpha=0.4)
    plt.tight_layout()
    
    # 5. 保存图片
    save_file = os.path.join(save_path, 'Monthly_Model_Weights_Evolution.png')
    plt.savefig(save_file, dpi=300, bbox_inches='tight') # bbox_inches 确保保存时不切掉图例
    
    print(f"📈 月频权重演化图已保存至: {save_file}")
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

# --- 1. 设置路径 (更新为 5 个模型) ---
paths = {
    'Ridge': r"../weights_diagnostic/paper_use/Ridge_MVO_Monthly_Weights.xlsx",
    'XGBoost': r"../weights_diagnostic/paper_use/XGB_Monthly_Weights.xlsx",
    'LightGBM': r"../weights_diagnostic/paper_use/LGBM_Monthly_Weights.xlsx",
    'Ensemble_Eqw': r"../weights_diagnostic/paper_use/Monthly_EW_Ensemble_Weights.xlsx",
    'Ensemble_CEnet': r"../weights_diagnostic/paper_use/HS300_CENet_Monthly_Final_Weights.xlsx"
}

# 读取行业映射数据
# 确保数据源包含 Indexcode 或 IndustryName
df_industry = pd.read_parquet('hs300_daily_data.parquet')
df_industry['TradingDate'] = pd.to_datetime(df_industry['TradingDate']).dt.normalize()
df_industry['Symbol'] = df_industry['Symbol'].astype(str).str.zfill(6)

def plot_five_models_with_industry_monthly(target_month, save_path='.'):
    """
    针对 5 个模型设计的月末持仓环形图分析
    target_month: 格式如 '2024-09'
    """
    if not os.path.exists(save_path):
        os.makedirs(save_path)

    # A. 确定该月的实际月末交易日
    available_dates = df_industry[df_industry['TradingDate'].dt.to_period('M') == target_month]['TradingDate'].unique()
    if len(available_dates) == 0:
        print(f"❌ 找不到该月的数据: {target_month}")
        return
    
    target_date = pd.to_datetime(max(available_dates)).normalize()
    print(f"📅 选定的月末分析基准日: {target_date.date()}")

    # 布局配置: 2行3列 (最后一块画布作为图例说明或留空)
    fig, axes = plt.subplots(2, 3, figsize=(24, 14))
    axes = axes.flatten()
    
    # 提取目标日期的行业映射
    industry_snap = df_industry[df_industry['TradingDate'] == target_date].set_index('Symbol')['Indexcode'].to_dict()
    
    model_names = list(paths.keys())

    for i in range(len(axes)):
        # 如果超出模型数量，隐藏坐标轴并显示统计信息
        if i >= len(model_names):
            axes[i].axis('off')
            if i == 5: # 在最后一个位置添加文字说明
                axes[i].text(0.5, 0.5, 
                            f"Analysis Summary:\n\nUniverse: HS300\nDate: {target_date.date()}\nModels: {len(model_names)}\nOptimization: MVO", 
                            ha='center', va='center', fontsize=16, 
                            fontweight='bold', bbox=dict(facecolor='none', edgecolor='gray', pad=10))
            continue

        name = model_names[i]
        path = paths[name]

        if not os.path.exists(path):
            axes[i].text(0.5, 0.5, f'File Not Found:\n{name}', ha='center', va='center', color='red', fontsize=14)
            axes[i].set_title(name, fontsize=20, fontweight='bold')
            continue

        # B. 读取权重
        df = pd.read_excel(path)
        df['date'] = pd.to_datetime(df['date']).dt.normalize()
        df['symbol'] = df['symbol'].astype(str).str.zfill(6)
        
        # C. 提取目标日持仓
        day_data = df[df['date'] == target_date].sort_values('weight', ascending=False)
        
        # 容错处理：如果当天没数据，找本月内最近的数据
        if day_data.empty:
            last_avail = df[df['date'].dt.to_period('M') == target_month]['date'].max()
            if pd.isna(last_avail):
                axes[i].text(0.5, 0.5, f'No data for {name}\nin {target_month}', ha='center', va='center', fontsize=14)
                axes[i].set_title(name, fontsize=20, fontweight='bold')
                continue
            day_data = df[df['date'] == last_avail].sort_values('weight', ascending=False)

        # D. 匹配行业与整理前 10 大持仓
        day_data['Indexcode'] = day_data['symbol'].map(industry_snap).fillna('Other/Suspended')
        
        top_n = 10
        plot_df = day_data.head(top_n).copy()
        plot_df['label'] = plot_df.apply(lambda r: f"{r['symbol']}\n{r['Indexcode']}", axis=1)
        
        others_weight = day_data.iloc[top_n:]['weight'].sum()
        if others_weight > 0:
            others_row = pd.DataFrame({'label': ['Others'], 'weight': [others_weight]})
            plot_df = pd.concat([plot_df, others_row], ignore_index=True)
        
        # E. 绘图
        wedges, texts, autotexts = axes[i].pie(
            plot_df['weight'], 
            labels=plot_df['label'], 
            autopct='%1.1f%%', 
            startangle=140,
            pctdistance=0.8, 
            colors=plt.cm.Spectral(np.linspace(0, 1, len(plot_df))), 
            wedgeprops={'width': 0.4, 'edgecolor': 'w'},
            labeldistance=1.1
        )
        
        axes[i].set_title(f'{name}', fontsize=22, fontweight='bold', pad=15)
        plt.setp(autotexts, size=10, weight="bold")
        plt.setp(texts, size=10)

    # 整体大标题
    plt.suptitle(f'Monthly Portfolio Allocation Breakdown: {target_month} (Reference Date: {target_date.date()})', 
                 fontsize=30, y=1.02, fontweight='bold')
    
    plt.tight_layout()
    
    # 保存图片
    save_file = os.path.join(save_path, f'Monthly_5_Model_Allocation_{target_month}.png')
    plt.savefig(save_file, dpi=300, bbox_inches='tight')
    print(f"🚀 分析图已成功生成: {save_file}")
    plt.show()

# --- 执行 ---
# 分析 2024年9月
plot_five_models_with_industry_monthly('2024-10')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

# --- 1. 配置路径 ---
paths = {
    'Ridge': r"../weights_diagnostic/paper_use/Ridge_MVO_Monthly_Weights.xlsx",
    'XGBoost': r"../weights_diagnostic/paper_use/XGB_Monthly_Weights.xlsx",
    'LightGBM': r"../weights_diagnostic/paper_use/LGBM_Monthly_Weights.xlsx",
    'Ensemble_Eqw': r"../weights_diagnostic/paper_use/Monthly_EW_Ensemble_Weights.xlsx",
    'Ensemble_CEnet': r"../weights_diagnostic/paper_use/HS300_CENet_Monthly_Final_Weights.xlsx"
}

# 读取行情数据
df_raw = pd.read_parquet('df_factor_monthly.parquet')
df_raw['TradingDate'] = pd.to_datetime(df_raw['TradingDate']).dt.normalize()
df_raw['Symbol'] = df_raw['Symbol'].astype(str).str.zfill(6)

def compare_next_month_performance(setup_month, forward_month, save_path='.'):
    """
    setup_month: '2024-09' (在此月末确定的持仓)
    forward_month: '2024-10' (观察此月份的收益)
    """
    # A. 提取 10 月份的真实收益率 (作为标签)
    # 在量化数据中，10月的 target 应该是 10月最后一个交易日对应的全月收益
    month_data = df_raw[df_raw['TradingDate'].dt.to_period('M') == forward_month]
    if month_data.empty:
        print(f"❌ 找不到目标收益月份的数据: {forward_month}")
        return

    last_date_of_forward = month_data['TradingDate'].max()
    ret_snap = month_data[month_data['TradingDate'] == last_date_of_forward].set_index('Symbol')['target_monthly_ret_y'].to_dict()
    
    results = []

    # B. 遍历模型计算 9 月末持仓在 10 月的收益
    for name, path in paths.items():
        if not os.path.exists(path): continue
        
        df_w = pd.read_excel(path)
        df_w['date'] = pd.to_datetime(df_w['date']).dt.normalize()
        df_w['symbol'] = df_w['symbol'].astype(str).str.zfill(6)
        
        # 锁定 9 月最后一个交易日的权重
        setup_dates = df_w[df_w['date'].dt.to_period('M') == setup_month]['date'].unique()
        if len(setup_dates) == 0: continue
        last_setup_date = max(setup_dates)
        
        day_weight = df_w[df_w['date'] == last_setup_date]
        
        # 计算持仓在 10 月实现的收益
        day_weight['forward_ret'] = day_weight['symbol'].map(ret_snap).fillna(0)
        port_ret = (day_weight['weight'] * day_weight['forward_ret']).sum()
        
        results.append({'Model': name, 'Return': port_ret})

    # C. 基准对比 (HS300 在 10 月的收益)
    bench_ret = np.mean(list(ret_snap.values()))
    results.append({'Model': 'HS300_Benchmark', 'Return': bench_ret})

    # D. 绘图
    res_df = pd.DataFrame(results).sort_values('Return', ascending=False)
    plt.figure(figsize=(13, 7))
    
    # 颜色区分：基准用灰色，最强模型用橙色，其他用蓝色
    colors = []
    max_ret = res_df['Return'].max()
    for _, row in res_df.iterrows():
        if row['Model'] == 'HS300_Benchmark': colors.append('#7f7f7f')
        elif row['Return'] == max_ret: colors.append('#d62728') # 红色高亮冠军
        else: colors.append('#1f77b4')

    bars = plt.bar(res_df['Model'], res_df['Return'] * 100, color=colors, alpha=0.8, edgecolor='black')
    
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval + (0.2 if yval>0 else -0.5), 
                 f'{yval:.2f}%', ha='center', va='bottom', fontweight='bold', fontsize=12)

    plt.axhline(0, color='black', lw=1.5)
    plt.title(f'Holding Performance: {setup_month} Setup vs {forward_month} Realized Return', fontsize=18, fontweight='bold', pad=25)
    plt.ylabel('Realized Monthly Return (%)', fontsize=14)
    plt.grid(axis='y', linestyle=':', alpha=0.7)
    
    save_file = os.path.join(save_path, f'Forward_Return_Analysis_{forward_month}.png')
    plt.savefig(save_file, dpi=300, bbox_inches='tight')
    plt.show()

# --- 执行 ---
# 比较 10 月末的持仓在 11 月的表现
compare_next_month_performance('2024-10', '2024-11')
```
