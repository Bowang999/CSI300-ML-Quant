```python
import pandas as pd
import os
from pathlib import Path

# ==========================================
# 1. 路径配置 (请修改为你的实际文件夹路径)
# ==========================================
# 权重文件夹：存放类似 "S_CON_WINDCODE" 的原始文件
WEIGHT_RAW_DIR = r"../RAW_DATA/权重原始文件夹"
# 指数收盘价文件夹：存放类似 "日期, 收盘价" 的原始文件
INDEX_RAW_DIR = r"../RAW_DATA/指数收盘价原始文件夹"

# 清洗后存储的文件夹
CLEAN_SAVE_DIR = r"../RAW_DATA/Clean_Data"
os.makedirs(CLEAN_SAVE_DIR, exist_ok=True)

# ==========================================
# 2. 核心清洗函数
# ==========================================

def clean_weight_files(input_dir, output_dir):
    """批量清洗权重文件"""
    processed_paths = []
    files = [f for f in os.listdir(input_dir) if f.endswith(('.xlsx', '.xls'))]
    
    for file in files:
        full_path = os.path.join(input_dir, file)
        print(f"正在清洗权重文件: {file}...")
        
        # 读取数据 (Sheet1)
        df = pd.read_excel(full_path)
        
        # 清洗逻辑
        # a. 股票代码: 000001.SZ -> 000001
        df['symbol'] = df['S_CON_WINDCODE'].astype(str).str.split('.').str[0].str.zfill(6)
        # b. 日期转换
        df['date'] = pd.to_datetime(df['TRADE_DT']).dt.normalize()
        # c. 权重转换: 1.46 -> 0.0146
        df['weight'] = df['I_WEIGHT'] / 100.0
        
        # 导出
        save_name = f"clean_weight_{Path(file).stem}.xlsx"
        save_path = os.path.join(output_dir, save_name)
        df[['date', 'symbol', 'weight']].sort_values(['date', 'symbol']).to_excel(save_path, index=False)
        processed_paths.append(save_path)
        
    return processed_paths

def clean_index_files(input_dir, output_dir):
    """批量清洗指数收盘价文件"""
    processed_paths = []
    files = [f for f in os.listdir(input_dir) if f.endswith(('.xlsx', '.xls'))]
    
    for file in files:
        full_path = os.path.join(input_dir, file)
        print(f"正在清洗指数文件: {file}...")
        
        # 读取数据 (假设两列: 日期, 收盘价)
        df = pd.read_excel(full_path)
        df.columns = ['date', 'close'] # 强制统一列名
        
        # 清洗逻辑
        df['date'] = pd.to_datetime(df['date']).dt.normalize()
        df = df.sort_values('date')
        # 计算日收益率
        df['index_ret'] = df['close'].pct_change()
        
        # 导出
        save_name = f"clean_index_{Path(file).stem}.xlsx"
        save_path = os.path.join(output_dir, save_name)
        df[['date', 'index_ret']].dropna().to_excel(save_path, index=False)
        processed_paths.append(save_path)
        
    return processed_paths

# ==========================================
# 3. 执行并输出地址
# ==========================================
if __name__ == "__main__":
    print("--- 启动自动化数据清洗流程 ---")
    
    weight_addresses = clean_weight_files(WEIGHT_RAW_DIR, CLEAN_SAVE_DIR)
    index_addresses = clean_index_files(INDEX_RAW_DIR, CLEAN_SAVE_DIR)
    
    print("\n" + "="*50)
    print("清洗完成！请复制以下地址用于回测：")
    print("="*50)
    
    print("\n[权重文件地址清单]:")
    for addr in weight_addresses:
        print(f"'{addr}'")
        
    print("\n[指数文件地址清单]:")
    for addr in index_addresses:
        print(f"'{addr}'")
```
