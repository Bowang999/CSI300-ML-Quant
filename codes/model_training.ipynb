```python
def numpy_spearmanr(x, y):
    """
    修改版：返回相关系数矩阵，以适配外部的 [0, 1] 索引
    """
    x = np.asarray(x)
    y = np.asarray(y)
    
    # 剔除 NaN
    mask = ~np.isnan(x) & ~np.isnan(y)
    x, y = x[mask], y[mask]

    # 如果样本太少或存在常数序列，返回一个全 0 矩阵防止外部索引报错
    if len(x) < 2 or np.std(x) == 0 or np.std(y) == 0:
        return np.zeros((2, 2))

    # 计算秩 (Ranks)
    def get_rank(arr):
        return arr.argsort().argsort() + 1.0

    rank_x = get_rank(x)
    rank_y = get_rank(y)

    # --- 核心修改：直接返回矩阵对象 ---
    # 不要在这里取 [0, 1]，让外部的第 138 行去取
    return np.corrcoef(rank_x, rank_y)
```

```python
import pandas as pd
import numpy as np
import cvxpy as cp
from sklearn.linear_model import RidgeCV
from sklearn.covariance import LedoitWolf
from tqdm import tqdm
import warnings

warnings.filterwarnings("ignore")

# --- 1. 配置与数据准备 ---
target = 'target_log_ret' 
features = [col for col in df_factor.columns if col.startswith('factor_')]
oos_start, oos_end = '2013-06-30', '2024-12-31'
lookback_window = 750 
train_step = 10    
pm_max_window = 750    # 最大回溯 750 天
risk_window_len = 120 

df_factor['TradingDate'] = pd.to_datetime(df_factor['TradingDate'])
all_dates = sorted(df_factor['TradingDate'].unique())
rebalance_dates = [d for d in all_dates if pd.to_datetime(oos_start) <= d <= pd.to_datetime(oos_end)]
df_lookup = {d: group for d, group in df_factor.groupby('TradingDate')}

# --- 2. 状态存储变量 ---
weight_records = []
performance_log = [] 

oos_num_raw = []
oos_num_smooth = []
oos_denominator_list = [] 

last_w_val = None 
last_mu_z_series = None     
last_mu_raw_series = None   
current_alpha = 100.0
model = None

print(f"🚀 启动 Ridge-MVO 整合策略 (动态个股均值基准, Cov: 120d)...")

# --- 3. 滚动预测与优化循环 ---
for i in tqdm(range(len(rebalance_dates))):
    date = rebalance_dates[i]
    date_idx = all_dates.index(date)
    
    # 基础保护：确保至少有足够数据训练初始模型
    if date_idx < lookback_window: continue
    
    # --- 3.1 动态再训练 ---
    if i % train_step == 0 or model is None:
        train_dates = all_dates[date_idx - lookback_window : date_idx]
        train_df = pd.concat([df_lookup[d] for d in train_dates if d in df_lookup])[[target] + features].dropna()
        X_train = train_df[features]
        y_train = train_df[target]
        model = RidgeCV(alphas=[0.1, 1.0, 10.0, 100.0, 1000.0], scoring='neg_mean_squared_error')
        model.fit(X_train, y_train)
        current_alpha = model.alpha_

    # --- 3.2 样本外预测与信号加工 ---
    current_day = df_lookup.get(date)
    if current_day is None: continue
    current_day = current_day.dropna(subset=features).copy()
    
    # 岭回归原始预测
    mu_raw_output = model.predict(current_day[features])
    
    # MVO 用的 Z-score 加工
    mu_z_raw_val = (mu_raw_output - np.mean(mu_raw_output)) / (np.std(mu_raw_output) + 1e-8)
    mu_z_raw_val = np.clip(mu_z_raw_val, -3, 3) 
    
    # 平滑逻辑
    if last_mu_z_series is not None:
        mu_raw_smooth_val = 0.3 * mu_raw_output + 0.7 * last_mu_raw_series.reindex(current_day['Symbol']).fillna(0).values
        mu_z_smooth_val = 0.3 * mu_z_raw_val + 0.7 * last_mu_z_series.reindex(current_day['Symbol']).fillna(0).values
    else:
        mu_raw_smooth_val = mu_raw_output
        mu_z_smooth_val = mu_z_raw_val
    
    last_mu_raw_series = pd.Series(mu_raw_smooth_val, index=current_day['Symbol'].tolist())
    last_mu_z_series = pd.Series(mu_z_smooth_val, index=current_day['Symbol'].tolist())

    # --- 3.3 MVO 资产配置 ---
    trade_candidates = current_day[current_day['is_hs300'] == True]['Symbol'].tolist()
    
    # 风险窗口 (固定 120d)
    risk_dates = all_dates[date_idx - risk_window_len : date_idx]
    risk_hist_df = pd.concat([df_lookup[d][['TradingDate', 'Symbol', target]] for d in risk_dates if d in df_lookup])
    
    # 【核心修改】：动态 PM 窗口 (Expanding up to 750d)
    # 取当前日期前所有的可用历史，但最多取 750 天
    actual_pm_start_idx = max(0, date_idx - pm_max_window)
    pm_dates = all_dates[actual_pm_start_idx : date_idx]
    pm_hist_df = pd.concat([df_lookup[d][['Symbol', target]] for d in pm_dates if d in df_lookup])
    
    # 直接按个股分组求均值。如果个股只有 3 天数据，均值就是 3 天的；如果有 750 天，就是 750 天的。
    rolling_pm_series = pm_hist_df.groupby('Symbol')[target].mean()
    
    v_matrix = risk_hist_df[risk_hist_df['Symbol'].isin(trade_candidates)].pivot(index='TradingDate', columns='Symbol', values=target).fillna(0)
    active_symbols = [s for s in trade_candidates if s in v_matrix.columns]
    
    if len(active_symbols) < 50: continue 

    # MVO 优化
    mu_for_mvo = last_mu_z_series.loc[active_symbols].values * 0.005 
    Sigma = LedoitWolf().fit(v_matrix[active_symbols].values).covariance_
    Sigma = (Sigma + Sigma.T) / 2
    
    n = len(active_symbols)
    w = cp.Variable(n)
    curr_last_w = last_w_val.reindex(active_symbols).fillna(0).values if last_w_val is not None else np.ones(n)/n
    
    gamma, tc_coeff = 3.0, 0.015 
    obj = cp.Maximize(mu_for_mvo @ w - 0.5 * gamma * cp.quad_form(w, Sigma) - tc_coeff * cp.norm(w - curr_last_w, 1))
    constraints = [cp.sum(w) == 1, w >= 0, w <= 1] 
    
    try:
        prob = cp.Problem(obj, constraints)
        prob.solve(solver=cp.OSQP, eps_abs=1e-5, eps_rel=1e-5)
        weights = w.value
        if weights is None: raise ValueError()
        weights[weights < 1e-4] = 0; weights /= weights.sum()
    except:
        weights = curr_last_w
        
    last_w_val = pd.Series(weights, index=active_symbols)
    weight_records.append(pd.DataFrame({'date': date, 'symbol': active_symbols, 'weight': weights}))
    
    # --- 3.4 统计诊断 (严格使用个股动态均值作为基准) ---
    if target in current_day.columns:
        y_real = current_day.set_index('Symbol').reindex(active_symbols)[target].values
        
        # 提取个股自身的动态 PM。如果某些极端情况缺失，则设为 0 (不使用截面均值兜底)
        daily_pm_bench = rolling_pm_series.reindex(active_symbols).fillna(0).values

        # 初始信号误差 (Raw)
        pred_raw = pd.Series(mu_raw_output, index=current_day['Symbol']).reindex(active_symbols).values
        err_sq_raw = np.sum((y_real - pred_raw)**2)
        ic_raw = numpy_spearmanr(pred_raw, y_real)[0, 1] if len(y_real) > 1 else 0

        # 平滑信号误差 (Smooth)
        pred_smooth = last_mu_raw_series.loc[active_symbols].values
        err_sq_smooth = np.sum((y_real - pred_smooth)**2)
        ic_smooth = numpy_spearmanr(pred_smooth, y_real)[0, 1] if len(y_real) > 1 else 0
        
        # 个股历史均值基准误差
        bench_err_sq = np.sum((y_real - daily_pm_bench)**2)

        oos_num_raw.append(err_sq_raw)
        oos_num_smooth.append(err_sq_smooth)
        oos_denominator_list.append(bench_err_sq)

        performance_log.append({
            'Date': date, 
            'IC_Raw': ic_raw,
            'IC_Smooth': ic_smooth,
            'R2_OS_Day_Raw': 1 - (err_sq_raw / (bench_err_sq + 1e-8)),
            'R2_OS_Day_Smooth': 1 - (err_sq_smooth / (bench_err_sq + 1e-8)),
            'Turnover': np.sum(np.abs(weights - curr_last_w)),
            'Alpha_Used': current_alpha
        })

# --- 4. 统计分析与导出 ---
if weight_records:
    total_bench = sum(oos_denominator_list) + 1e-8
    final_r2_raw = 1 - (sum(oos_num_raw) / total_bench)
    final_r2_smooth = 1 - (sum(oos_num_smooth) / total_bench)
    
    perf_df = pd.DataFrame(performance_log)
    print("\n" + "="*40)
    print(f"📊 Ridge-MVO 终极诊断表 (个股动态 PM 基准):")
    print(f"全样本 $R^2_{{OS}}$ (初始信号): {final_r2_raw:.6f}")
    print(f"全样本 $R^2_{{OS}}$ (平滑信号): {final_r2_smooth:.6f}")
    print(f"平均 IC (初始信号): {perf_df['IC_Raw'].mean():.4f}")
    print(f"平均 IC (平滑信号): {perf_df['IC_Smooth'].mean():.4f}")
    print(f"平均换手率: {perf_df['Turnover'].mean():.4f}")
    print("="*40)
    
    perf_df.to_excel('Ridge_MVO_DynamicPM_Diagnostics.xlsx', index=False)
    pd.concat(weight_records)[lambda x: x['weight'] > 0].to_excel('Ridge_MVO_DynamicPM_Weights.xlsx', index=False)
```

```python
import pandas as pd
import numpy as np
import cvxpy as cp
import xgboost as xgb
from sklearn.covariance import LedoitWolf
from sklearn.model_selection import TimeSeriesSplit, GridSearchCV
from tqdm import tqdm
import warnings
import gc

warnings.filterwarnings("ignore")

# --- 1. 配置 ---
target = 'target_log_ret' 
features = [col for col in df_factor.columns if col.startswith('factor_')]
oos_start, oos_end = '2013-06-30', '2024-12-31'
lookback_window = 750 
train_step = 10
pm_max_window = 750    # 最大回溯 750 天计算个股均值
risk_window_len = 120  # 协方差窗口

df_factor['TradingDate'] = pd.to_datetime(df_factor['TradingDate'])
all_dates = sorted(df_factor['TradingDate'].unique())
rebalance_dates = [d for d in all_dates if pd.to_datetime(oos_start) <= d <= pd.to_datetime(oos_end)]
df_lookup = {d: group for d, group in df_factor.groupby('TradingDate')}

# --- 2. XGBoost 参数空间 ---
base_params = {
    'learning_rate': 0.05,
    'n_estimators': 100,
    'colsample_bytree': 0.6,
    'reg_alpha': 1.0,
    'tree_method': 'hist',
    'device': 'cpu',  # 如有GPU可改为 'cuda'
    'n_jobs': -1,
    'verbosity': 0
}

param_grid = {
    'max_depth': [3, 5],
    'reg_lambda': [10.0, 50.0, 100.0]
}

# --- 3. 滚动回测变量 ---
weight_records = []
performance_log = [] 

# R2 核心监控组件 (全样本累加)
oos_num_raw = []
oos_num_smooth = []
oos_denominator_list = [] 

last_w_val = None 
last_mu_z_series = None     # 供 MVO 使用的平滑 Z-score
last_mu_raw_series = None   # 供 R2 评估使用的平滑原始量纲信号
current_best_params = {}
model = None

print(f"🚀 启动 XGBoost 整合策略 (动态个股 PM 基准, 严格逻辑对齐)...")

# --- 4. 滚动预测与优化循环 ---
for i in tqdm(range(len(rebalance_dates))):
    date = rebalance_dates[i]
    date_idx = all_dates.index(date)
    
    # 基础保护：确保至少有足够数据训练初始模型
    if date_idx < lookback_window: continue
    
    # --- 4.1 动态再训练 (每10天触发一次) ---
    if i % train_step == 0 or model is None:
        if 'model' in locals(): del model
        gc.collect()

        train_dates = all_dates[date_idx - lookback_window : date_idx]
        train_df = pd.concat([df_lookup[d] for d in train_dates if d in df_lookup])[[target] + features].dropna()
        
        X_train = train_df[features]
        # XGBoost 直接训练原始收益率
        y_train = train_df[target]
        
        tscv = TimeSeriesSplit(n_splits=3)
        grid_search = GridSearchCV(
            estimator=xgb.XGBRegressor(**base_params),
            param_grid=param_grid,
            cv=tscv,
            scoring='neg_mean_squared_error',
            n_jobs=1 
        )
        grid_search.fit(X_train, y_train)
        
        model = grid_search.best_estimator_
        current_best_params = grid_search.best_params_
        del train_df; gc.collect()

    # --- 4.2 样本外预测与信号加工 ---
    current_day = df_lookup.get(date)
    if current_day is None: continue
    current_day = current_day.dropna(subset=features).copy()
    
    # XGBoost 原始预测输出 (原始收益率量纲)
    mu_raw_output = model.predict(current_day[features])
    
    # 加工：截面标准化 (用于 MVO 相对排序)
    mu_z_raw_val = (mu_raw_output - np.mean(mu_raw_output)) / (np.std(mu_raw_output) + 1e-8)
    mu_z_raw_val = np.clip(mu_z_raw_val, -3, 3) 
    
    # 时间序列平滑：同时维护 原始量纲 和 Z-score 两套平滑信号
    if last_mu_z_series is not None:
        # 1. 原始量纲预测值平滑 (用于 R2_Smooth 和 IC_Smooth 评估)
        mu_raw_smooth_val = 0.3 * mu_raw_output + 0.7 * last_mu_raw_series.reindex(current_day['Symbol']).fillna(0).values
        # 2. Z-score 平滑 (用于 MVO 资产配置)
        mu_z_smooth_val = 0.3 * mu_z_raw_val + 0.7 * last_mu_z_series.reindex(current_day['Symbol']).fillna(0).values
    else:
        mu_raw_smooth_val = mu_raw_output
        mu_z_smooth_val = mu_z_raw_val
    
    # 更新状态变量
    last_mu_raw_series = pd.Series(mu_raw_smooth_val, index=current_day['Symbol'].tolist())
    last_mu_z_series = pd.Series(mu_z_smooth_val, index=current_day['Symbol'].tolist())

    # --- 4.3 MVO 资产配置 ---
    trade_candidates = current_day[current_day['is_hs300'] == True]['Symbol'].tolist()
    
    # 风险窗口 120d / PM 基准窗口 750d
    risk_dates = all_dates[date_idx - risk_window_len : date_idx]
    actual_pm_start_idx = max(0, date_idx - pm_max_window)
    pm_dates = all_dates[actual_pm_start_idx : date_idx]
    
    risk_hist_df = pd.concat([df_lookup[d][['TradingDate', 'Symbol', target]] for d in risk_dates if d in df_lookup])
    pm_hist_df = pd.concat([df_lookup[d][['Symbol', target]] for d in pm_dates if d in df_lookup])
    
    # 计算个股动态 PM (Prevailing Mean)
    rolling_pm_series = pm_hist_df.groupby('Symbol')[target].mean()
    
    v_matrix = risk_hist_df[risk_hist_df['Symbol'].isin(trade_candidates)].pivot(index='TradingDate', columns='Symbol', values=target).fillna(0)
    active_symbols = [s for s in trade_candidates if s in v_matrix.columns]
    
    if len(active_symbols) < 50: continue 

    # MVO 逻辑 (使用 Z-score 平滑信号映射预期收益)
    mu_for_mvo = last_mu_z_series.loc[active_symbols].values * 0.005 
    Sigma = LedoitWolf().fit(v_matrix[active_symbols].values).covariance_
    Sigma = (Sigma + Sigma.T) / 2
    
    n = len(active_symbols)
    w = cp.Variable(n)
    curr_last_w = last_w_val.reindex(active_symbols).fillna(0).values if last_w_val is not None else np.ones(n)/n
    
    gamma, tc_coeff = 3.0, 0.015 
    obj = cp.Maximize(mu_for_mvo @ w - 0.5 * gamma * cp.quad_form(w, Sigma) - tc_coeff * cp.norm(w - curr_last_w, 1))
    constraints = [cp.sum(w) == 1, w >= 0, w <= 1] 
    
    try:
        prob = cp.Problem(obj, constraints)
        prob.solve(solver=cp.OSQP, eps_abs=1e-5, eps_rel=1e-5)
        weights = w.value
        if weights is None: raise ValueError()
        weights[weights < 1e-4] = 0; weights /= weights.sum()
    except:
        weights = curr_last_w
        
    last_w_val = pd.Series(weights, index=active_symbols)
    weight_records.append(pd.DataFrame({'date': date, 'symbol': active_symbols, 'weight': weights}))
    
    # --- 4.4 统计诊断 (严格计算原始量纲 IC/R2) ---
    if target in current_day.columns:
        y_real = current_day.set_index('Symbol').reindex(active_symbols)[target].values
        # 获取个股动态 PM 基准
        daily_pm_bench = rolling_pm_series.reindex(active_symbols).fillna(0).values

        # A. 初始信号评估 (XGB 直接预测值)
        pred_raw = pd.Series(mu_raw_output, index=current_day['Symbol']).reindex(active_symbols).values
        err_sq_raw = np.sum((y_real - pred_raw)**2)
        ic_raw = numpy_spearmanr(pred_raw, y_real)[0, 1] if len(y_real) > 1 else 0

        # B. 平滑信号评估 (平滑后的原始量纲预测值)
        pred_smooth = last_mu_raw_series.loc[active_symbols].values
        err_sq_smooth = np.sum((y_real - pred_smooth)**2)
        ic_smooth = numpy_spearmanr(pred_smooth, y_real)[0, 1] if len(y_real) > 1 else 0
        
        # C. 基准误差
        bench_err_sq = np.sum((y_real - daily_pm_bench)**2)

        oos_num_raw.append(err_sq_raw)
        oos_num_smooth.append(err_sq_smooth)
        oos_denominator_list.append(bench_err_sq)

        performance_log.append({
            'Date': date, 
            'IC_Raw': ic_raw,
            'IC_Smooth': ic_smooth,
            'R2_OS_Day_Raw': 1 - (err_sq_raw / (bench_err_sq + 1e-8)),
            'R2_OS_Day_Smooth': 1 - (err_sq_smooth / (bench_err_sq + 1e-8)),
            'Turnover': np.sum(np.abs(weights - curr_last_w)),
            'Params': str(current_best_params)
        })

# --- 5. 汇总导出 ---
if weight_records:
    total_bench = sum(oos_denominator_list) + 1e-8
    final_r2_raw = 1 - (sum(oos_num_raw) / total_bench)
    final_r2_smooth = 1 - (sum(oos_num_smooth) / total_bench)
    
    perf_df = pd.DataFrame(performance_log)
    print("\n" + "="*45)
    print(f"📊 XGBoost 回测诊断表 (个股动态 PM 基准):")
    print(f"全样本 $R^2_{{OS}}$ (初始信号): {final_r2_raw:.6f}")
    print(f"全样本 $R^2_{{OS}}$ (平滑信号): {final_r2_smooth:.6f}")
    print(f"平均 IC (初始信号): {perf_df['IC_Raw'].mean():.4f}")
    print(f"平均 IC (平滑信号): {perf_df['IC_Smooth'].mean():.4f}")
    print(f"平均换手率: {perf_df['Turnover'].mean():.4f}")
    print("="*45)
    
    perf_df.to_excel('XGB_DynamicPM_Diagnostics.xlsx', index=False)
    pd.concat(weight_records)[lambda x: x['weight'] > 0].to_excel('XGB_DynamicPM_Weights.xlsx', index=False)
```

```python
import pandas as pd
import numpy as np
import cvxpy as cp
import lightgbm as lgb
from sklearn.covariance import LedoitWolf
from sklearn.model_selection import TimeSeriesSplit, GridSearchCV
from tqdm import tqdm
import warnings
import gc

warnings.filterwarnings("ignore")

# --- 1. 配置 ---
target = 'target_log_ret' 
features = [col for col in df_factor.columns if col.startswith('factor_')]
oos_start, oos_end = '2013-06-30', '2024-12-31'
lookback_window = 750 
train_step = 10
pm_max_window = 750    # 最大回溯 750 天计算个股均值
risk_window_len = 120  # 协方差窗口

df_factor['TradingDate'] = pd.to_datetime(df_factor['TradingDate'])
all_dates = sorted(df_factor['TradingDate'].unique())
rebalance_dates = [d for d in all_dates if pd.to_datetime(oos_start) <= d <= pd.to_datetime(oos_end)]
df_lookup = {d: group for d, group in df_factor.groupby('TradingDate')}

# --- 2. LightGBM 动态搜索空间配置 ---
base_params = {
    'n_estimators': 100,
    'learning_rate': 0.05,
    'feature_fraction': 0.6,
    'lambda_l1': 1.0,
    'verbosity': -1,
    'n_jobs': -1
}

param_grid = {
    'max_depth': [3, 5],
    'lambda_l2': [10.0, 50.0, 100.0]
}

# --- 3. 稳健滚动回测变量 ---
weight_records = []
performance_log = [] 

# R2 核心监控组件 (全样本累加)
oos_num_raw = []
oos_num_smooth = []
oos_denominator_list = [] 

last_w_val = None 
last_mu_z_series = None     # 供 MVO 使用的平滑 Z-score
last_mu_raw_series = None   # 供 R2 评估使用的平滑原始量纲信号
current_best_params = {}
model = None

print(f"🚀 启动 LightGBM 整合策略 (动态个股 PM 基准, 严格逻辑对齐)...")

# --- 4. 滚动预测与优化循环 ---
for i in tqdm(range(len(rebalance_dates))):
    date = rebalance_dates[i]
    date_idx = all_dates.index(date)
    
    # 基础保护：确保至少有足够数据训练初始模型
    if date_idx < lookback_window: continue
    
    # --- 4.1 动态再训练 (每10天触发一次) ---
    if i % train_step == 0 or model is None:
        if 'model' in locals(): del model
        gc.collect()

        train_dates = all_dates[date_idx - lookback_window : date_idx]
        train_df = pd.concat([df_lookup[d] for d in train_dates if d in df_lookup])[[target] + features].dropna()
        
        X_train = train_df[features]
        # LightGBM 直接训练原始收益率
        y_train = train_df[target]
        
        tscv = TimeSeriesSplit(n_splits=3)
        grid_search = GridSearchCV(
            estimator=lgb.LGBMRegressor(**base_params),
            param_grid=param_grid,
            cv=tscv,
            scoring='neg_mean_squared_error',
            n_jobs=1 
        )
        grid_search.fit(X_train, y_train)
        
        model = grid_search.best_estimator_
        current_best_params = grid_search.best_params_
        del train_df; gc.collect()

    # --- 4.2 样本外预测与信号加工 ---
    current_day = df_lookup.get(date)
    if current_day is None: continue
    current_day = current_day.dropna(subset=features).copy()
    
    # LightGBM 原始预测输出 (原始收益率量纲)
    mu_raw_output = model.predict(current_day[features])
    
    # 加工：截面标准化 (用于 MVO 相对排序)
    mu_z_raw_val = (mu_raw_output - np.mean(mu_raw_output)) / (np.std(mu_raw_output) + 1e-8)
    mu_z_raw_val = np.clip(mu_z_raw_val, -3, 3) 
    
    # 时间序列平滑：同时维护 原始量纲 和 Z-score 两套平滑信号
    if last_mu_z_series is not None:
        # 1. 原始量纲预测值平滑 (用于 R2_Smooth 和 IC_Smooth 评估)
        mu_raw_smooth_val = 0.3 * mu_raw_output + 0.7 * last_mu_raw_series.reindex(current_day['Symbol']).fillna(0).values
        # 2. Z-score 平滑 (用于 MVO 资产配置)
        mu_z_smooth_val = 0.3 * mu_z_raw_val + 0.7 * last_mu_z_series.reindex(current_day['Symbol']).fillna(0).values
    else:
        mu_raw_smooth_val = mu_raw_output
        mu_z_smooth_val = mu_z_raw_val
    
    # 更新状态变量
    last_mu_raw_series = pd.Series(mu_raw_smooth_val, index=current_day['Symbol'].tolist())
    last_mu_z_series = pd.Series(mu_z_smooth_val, index=current_day['Symbol'].tolist())

    # --- 4.3 MVO 资产配置 ---
    trade_candidates = current_day[current_day['is_hs300'] == True]['Symbol'].tolist()
    
    # 风险窗口 120d / PM 基准窗口 750d
    risk_dates = all_dates[date_idx - risk_window_len : date_idx]
    actual_pm_start_idx = max(0, date_idx - pm_max_window)
    pm_dates = all_dates[actual_pm_start_idx : date_idx]
    
    risk_hist_df = pd.concat([df_lookup[d][['TradingDate', 'Symbol', target]] for d in risk_dates if d in df_lookup])
    pm_hist_df = pd.concat([df_lookup[d][['Symbol', target]] for d in pm_dates if d in df_lookup])
    
    # 计算个股动态 PM (Prevailing Mean)
    rolling_pm_series = pm_hist_df.groupby('Symbol')[target].mean()
    
    v_matrix = risk_hist_df[risk_hist_df['Symbol'].isin(trade_candidates)].pivot(index='TradingDate', columns='Symbol', values=target).fillna(0)
    active_symbols = [s for s in trade_candidates if s in v_matrix.columns]
    
    if len(active_symbols) < 50: continue 

    # MVO 逻辑 (使用 Z-score 平滑信号映射预期收益)
    mu_for_mvo = last_mu_z_series.loc[active_symbols].values * 0.005 
    Sigma = LedoitWolf().fit(v_matrix[active_symbols].values).covariance_
    Sigma = (Sigma + Sigma.T) / 2
    
    n = len(active_symbols)
    w = cp.Variable(n)
    curr_last_w = last_w_val.reindex(active_symbols).fillna(0).values if last_w_val is not None else np.ones(n)/n
    
    gamma, tc_coeff = 3.0, 0.015 
    obj = cp.Maximize(mu_for_mvo @ w - 0.5 * gamma * cp.quad_form(w, Sigma) - tc_coeff * cp.norm(w - curr_last_w, 1))
    constraints = [cp.sum(w) == 1, w >= 0, w <= 1] 
    
    try:
        prob = cp.Problem(obj, constraints)
        prob.solve(solver=cp.OSQP, eps_abs=1e-5, eps_rel=1e-5)
        weights = w.value
        if weights is None: raise ValueError()
        weights[weights < 1e-4] = 0; weights /= weights.sum()
    except:
        weights = curr_last_w
        
    last_w_val = pd.Series(weights, index=active_symbols)
    weight_records.append(pd.DataFrame({'date': date, 'symbol': active_symbols, 'weight': weights}))
    
    # --- 4.4 统计诊断 (严格计算原始量纲 IC/R2) ---
    if target in current_day.columns:
        y_real = current_day.set_index('Symbol').reindex(active_symbols)[target].values
        # 获取个股动态 PM 基准
        daily_pm_bench = rolling_pm_series.reindex(active_symbols).fillna(0).values

        # A. 初始信号评估 (LGBM 直接预测值)
        pred_raw = pd.Series(mu_raw_output, index=current_day['Symbol']).reindex(active_symbols).values
        err_sq_raw = np.sum((y_real - pred_raw)**2)
        ic_raw = numpy_spearmanr(pred_raw, y_real)[0, 1] if len(y_real) > 1 else 0

        # B. 平滑信号评估 (平滑后的原始量纲预测值)
        pred_smooth = last_mu_raw_series.loc[active_symbols].values
        err_sq_smooth = np.sum((y_real - pred_smooth)**2)
        ic_smooth = numpy_spearmanr(pred_smooth, y_real)[0, 1] if len(y_real) > 1 else 0
        
        # C. 基准误差
        bench_err_sq = np.sum((y_real - daily_pm_bench)**2)

        oos_num_raw.append(err_sq_raw)
        oos_num_smooth.append(err_sq_smooth)
        oos_denominator_list.append(bench_err_sq)

        performance_log.append({
            'Date': date, 
            'IC_Raw': ic_raw,
            'IC_Smooth': ic_smooth,
            'R2_OS_Day_Raw': 1 - (err_sq_raw / (bench_err_sq + 1e-8)),
            'R2_OS_Day_Smooth': 1 - (err_sq_smooth / (bench_err_sq + 1e-8)),
            'Turnover': np.sum(np.abs(weights - curr_last_w)),
            'Best_Params': str(current_best_params)
        })

# --- 5. 汇总导出 ---
if weight_records:
    total_bench = sum(oos_denominator_list) + 1e-8
    final_r2_raw = 1 - (sum(oos_num_raw) / total_bench)
    final_r2_smooth = 1 - (sum(oos_num_smooth) / total_bench)
    
    perf_df = pd.DataFrame(performance_log)
    print("\n" + "="*45)
    print(f"📊 LightGBM 回测诊断表 (个股动态 PM 基准):")
    print(f"全样本 $R^2_{{OS}}$ (初始信号): {final_r2_raw:.6f}")
    print(f"全样本 $R^2_{{OS}}$ (平滑信号): {final_r2_smooth:.6f}")
    print(f"平均 IC (初始信号): {perf_df['IC_Raw'].mean():.4f}")
    print(f"平均 IC (平滑信号): {perf_df['IC_Smooth'].mean():.4f}")
    print(f"平均换手率: {perf_df['Turnover'].mean():.4f}")
    print("="*45)
    
    perf_df.to_excel('LGBM_DynamicPM_Diagnostics.xlsx', index=False)
    pd.concat(weight_records)[lambda x: x['weight'] > 0].to_excel('LGBM_DynamicPM_Weights.xlsx', index=False)
```

```python
import pandas as pd
import numpy as np
import cvxpy as cp
import lightgbm as lgb
import xgboost as xgb
from sklearn.linear_model import Ridge, ElasticNetCV
from sklearn.covariance import LedoitWolf
from tqdm import tqdm
import warnings
import gc

warnings.filterwarnings("ignore")

# --- 1. 配置 ---
target = 'target_log_ret' 
features = [col for col in df_factor.columns if col.startswith('factor_')]
oos_start, oos_end = '2013-06-30', '2024-12-31'
lookback_window = 750 
comb_window = 120 
train_step = 10
pm_max_window = 750    # 最大回溯 750 天计算个股均值
risk_window_len = 120  # 协方差窗口

df_factor['TradingDate'] = pd.to_datetime(df_factor['TradingDate'])
all_dates = sorted(df_factor['TradingDate'].unique())
rebalance_dates = [d for d in all_dates if pd.to_datetime(oos_start) <= d <= pd.to_datetime(oos_end)]
df_lookup = {d: group for d, group in df_factor.groupby('TradingDate')}

# --- 2. 锁定底层黄金参数 ---
params_lgb = {'n_estimators': 100, 'learning_rate': 0.01, 'max_depth': 3, 'lambda_l2': 50.0}
params_xgb = {'n_estimators': 100, 'learning_rate': 0.01, 'max_depth': 3, 'reg_lambda': 50.0}
params_ridge = {'alpha': 1000.0, 'random_state': 42}

# --- 3. 滚动回测变量 ---
weight_records = []
performance_log = [] 

# R2 核心监控组件 (全样本累加)
oos_num_raw = []
oos_num_smooth = []
oos_denominator_list = [] 

last_w_val = None 
last_mu_z_series = None     # 供 MVO 使用的平滑 Z-score 信号
last_mu_raw_series = None   # 供 R2 评估使用的平滑原始量纲信号
last_ensemble_weights = None 

print(f"🚀 启动 C-ENet 动态集成策略 (动态个股 PM 基准, 严格逻辑对齐)...")

# --- 4. 滚动预测与优化循环 ---
for i in tqdm(range(len(rebalance_dates))):
    date = rebalance_dates[i]
    date_idx = all_dates.index(date)
    
    if date_idx < lookback_window: continue
    
    # --- 4.1 动态再训练底层模型及集成权重 (每 10 天) ---
    if i % train_step == 0 or 'model_lgb' not in locals():
        train_dates = all_dates[date_idx - lookback_window : date_idx]
        train_df = pd.concat([df_lookup[d] for d in train_dates if d in df_lookup])[[target] + features].dropna()
        
        # A. 训练底层子模型 (基于原始收益率)
        model_lgb = lgb.LGBMRegressor(**params_lgb, verbosity=-1).fit(train_df[features], train_df[target])
        model_xgb = xgb.XGBRegressor(**params_xgb, verbosity=0).fit(train_df[features], train_df[target])
        model_ridge = Ridge(**params_ridge).fit(train_df[features], train_df[target])
        
        # B. C-ENet 动态集成权重计算
        comb_dates = train_dates[-comb_window:]
        comb_list = []
        for d in comb_dates:
            if d in df_lookup:
                day_data = df_lookup[d].dropna(subset=features)
                p1 = model_lgb.predict(day_data[features])
                p2 = model_xgb.predict(day_data[features])
                p3 = model_ridge.predict(day_data[features])
                y_true = day_data[target].values
                
                # 集成权重训练基于 Z-score 化的子预测与目标
                comb_list.append(pd.DataFrame({
                    'lgb': np.clip((p1 - p1.mean())/(p1.std()+1e-8), -3, 3),
                    'xgb': np.clip((p2 - p2.mean())/(p2.std()+1e-8), -3, 3),
                    'rid': np.clip((p3 - p3.mean())/(p3.std()+1e-8), -3, 3),
                    'target': np.clip((y_true - y_true.mean())/(y_true.std()+1e-8), -3, 3)
                }))
        
        c_df = pd.concat(comb_list).dropna()
        enet = ElasticNetCV(l1_ratio=[0.1, 0.5, 0.9], alphas=[0.01, 0.1, 1.0], 
                            positive=True, fit_intercept=False, cv=5)
        enet.fit(c_df[['lgb', 'xgb', 'rid']], c_df['target'])
        
        new_raw_weights = enet.coef_ / (enet.coef_.sum() + 1e-8) if enet.coef_.sum() > 1e-4 else np.array([1/3, 1/3, 1/3])
        
        # 权重动量平滑
        if last_ensemble_weights is None:
            current_ensemble_weights = new_raw_weights
        else:
            current_ensemble_weights = 0.9 * last_ensemble_weights + 0.1 * new_raw_weights
        last_ensemble_weights = current_ensemble_weights
        
        del train_df; gc.collect()

    current_day = df_lookup.get(date)
    if current_day is None: continue
    current_day = current_day.dropna(subset=features).copy()
    
    # --- 4.2 获取预测并进行加工 ---
    raw_preds_matrix = np.column_stack([
        model_lgb.predict(current_day[features]),
        model_xgb.predict(current_day[features]),
        model_ridge.predict(current_day[features])
    ])
    
    # A. 原始量纲合成信号 (评估用)
    mu_raw_output = raw_preds_matrix @ current_ensemble_weights
    
    # B. 截面标准化处理 (集成逻辑基于 Z-score 权重)
    preds_z = (raw_preds_matrix - raw_preds_matrix.mean(axis=0)) / (raw_preds_matrix.std(axis=0) + 1e-8)
    mu_z_raw_val = np.clip(preds_z @ current_ensemble_weights, -3, 3)
    
    # C. 时间序列平滑
    if last_mu_z_series is not None:
        mu_raw_smooth_val = 0.3 * mu_raw_output + 0.7 * last_mu_raw_series.reindex(current_day['Symbol']).fillna(0).values
        mu_z_smooth_val = 0.3 * mu_z_raw_val + 0.7 * last_mu_z_series.reindex(current_day['Symbol']).fillna(0).values
    else:
        mu_raw_smooth_val = mu_raw_output
        mu_z_smooth_val = mu_z_raw_val
    
    last_mu_raw_series = pd.Series(mu_raw_smooth_val, index=current_day['Symbol'].tolist())
    last_mu_z_series = pd.Series(mu_z_smooth_val, index=current_day['Symbol'].tolist())

    # --- 4.3 MVO 资产配置 ---
    trade_candidates = current_day[current_day['is_hs300'] == True]['Symbol'].tolist()
    
    risk_dates = all_dates[date_idx - risk_window_len : date_idx]
    actual_pm_start_idx = max(0, date_idx - pm_max_window)
    pm_dates = all_dates[actual_pm_start_idx : date_idx]
    
    risk_hist_df = pd.concat([df_lookup[d][['TradingDate', 'Symbol', target]] for d in risk_dates if d in df_lookup])
    pm_hist_df = pd.concat([df_lookup[d][['Symbol', target]] for d in pm_dates if d in df_lookup])
    
    rolling_pm_series = pm_hist_df.groupby('Symbol')[target].mean()
    
    v_matrix = risk_hist_df[risk_hist_df['Symbol'].isin(trade_candidates)].pivot(index='TradingDate', columns='Symbol', values=target).fillna(0)
    active_symbols = [s for s in trade_candidates if s in v_matrix.columns]
    
    if len(active_symbols) < 50: continue 

    # MVO 使用平滑后的 Z-score 信号
    mu_for_mvo = last_mu_z_series.loc[active_symbols].values * 0.005 
    Sigma = LedoitWolf().fit(v_matrix[active_symbols].values).covariance_
    Sigma = (Sigma + Sigma.T) / 2
    
    n = len(active_symbols); w = cp.Variable(n)
    curr_last_w = last_w_val.reindex(active_symbols).fillna(0).values if last_w_val is not None else np.ones(n)/n
    
    obj = cp.Maximize(mu_for_mvo @ w - 0.5 * 3.0 * cp.quad_form(w, Sigma) - 0.015 * cp.norm(w - curr_last_w, 1))
    constraints = [cp.sum(w) == 1, w >= 0, w <= 1]
    
    try:
        prob = cp.Problem(obj, constraints); prob.solve(solver=cp.OSQP)
        weights = w.value
        if weights is None: raise ValueError()
        weights[weights < 1e-4] = 0; weights /= weights.sum()
    except:
        weights = curr_last_w
        
    last_w_val = pd.Series(weights, index=active_symbols)
    weight_records.append(pd.DataFrame({'date': date, 'symbol': active_symbols, 'weight': weights}))
    
    # --- 4.4 统计诊断 (原始量纲评估) ---
    if target in current_day.columns:
        y_real = current_day.set_index('Symbol').reindex(active_symbols)[target].values
        daily_pm_bench = rolling_pm_series.reindex(active_symbols).fillna(0).values

        # A. 初始集成信号误差 (Raw)
        pred_raw = pd.Series(mu_raw_output, index=current_day['Symbol']).reindex(active_symbols).values
        err_sq_raw = np.sum((y_real - pred_raw)**2)
        ic_raw = numpy_spearmanr(pred_raw, y_real)[0, 1] if len(y_real) > 1 else 0

        # B. 平滑集成信号误差 (Smooth)
        pred_smooth = last_mu_raw_series.loc[active_symbols].values
        err_sq_smooth = np.sum((y_real - pred_smooth)**2)
        ic_smooth = numpy_spearmanr(pred_smooth, y_real)[0, 1] if len(y_real) > 1 else 0
        
        # C. 基准误差
        bench_err_sq = np.sum((y_real - daily_pm_bench)**2)

        oos_num_raw.append(err_sq_raw)
        oos_num_smooth.append(err_sq_smooth)
        oos_denominator_list.append(bench_err_sq)

        performance_log.append({
            'Date': date, 
            'IC_Raw': ic_raw,
            'IC_Smooth': ic_smooth,
            'R2_OS_Day_Raw': 1 - (err_sq_raw / (bench_err_sq + 1e-8)),
            'R2_OS_Day_Smooth': 1 - (err_sq_smooth / (bench_err_sq + 1e-8)),
            'Turnover': np.sum(np.abs(weights - curr_last_w)),
            'W_LGB': current_ensemble_weights[0],
            'W_XGB': current_ensemble_weights[1],
            'W_RID': current_ensemble_weights[2]
        })

# --- 5. 汇总导出 ---
if weight_records:
    pd.concat(weight_records)[lambda x: x['weight'] > 0].to_excel('HS300_CENet_Dynamic_Final.xlsx', index=False)
    
    perf_df = pd.DataFrame(performance_log)
    perf_df.to_excel('CENet_Dynamic_Diagnostics.xlsx', index=False)

    total_bench = sum(oos_denominator_list) + 1e-8
    final_r2_raw = 1 - (sum(oos_num_raw) / total_bench)
    final_r2_smooth = 1 - (sum(oos_num_smooth) / total_bench)
    
    print("\n" + "="*45)
    print(f"📊 C-ENet 动态集成统计结果 (个股动态 PM 基准):")
    print(f"全样本 $R^2_{{OS}}$ (初始信号): {final_r2_raw:.6f}")
    print(f"全样本 $R^2_{{OS}}$ (平滑信号): {final_r2_smooth:.6f}")
    print(f"平均 IC (初始信号): {perf_df['IC_Raw'].mean():.4f}")
    print(f"平均 IC (平滑信号): {perf_df['IC_Smooth'].mean():.4f}")
    print(f"平均换手率: {perf_df['Turnover'].mean():.4f}")
    print("="*45)
```

```python
import pandas as pd
import numpy as np
import cvxpy as cp
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from sklearn.covariance import LedoitWolf
from scipy.stats import skew, kurtosis
from tqdm import tqdm
import warnings
import copy

warnings.filterwarnings("ignore")

# --- 0. 高性能 AI 模型定义: LSTM + Vectorized VSN ---
class VSN(nn.Module):
    def __init__(self, input_dim, hidden_dim):
        super(VSN, self).__init__()
        self.weight_net = nn.Sequential(
            nn.Linear(input_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, input_dim),
            nn.Softmax(dim=-1)
        )
        
    def forward(self, x):
        # 基于序列最后一个时间点计算特征权重
        weights = self.weight_net(x[:, -1, :]) 
        return x * weights.unsqueeze(1)

class LSTM_VSN_Model(nn.Module):
    def __init__(self, input_dim, hidden_dim):
        super(LSTM_VSN_Model, self).__init__()
        self.vsn = VSN(input_dim, hidden_dim)
        self.lstm = nn.LSTM(input_dim, hidden_dim, batch_first=True)
        self.fc = nn.Linear(hidden_dim, 1)
        
    def forward(self, x):
        x = self.vsn(x)
        _, (h_n, _) = self.lstm(x)
        return self.fc(h_n[-1])

# --- 1. 配置与数据准备 ---
target = 'target_log_ret' 
features = [col for col in df_factor.columns if col.startswith('factor_')]
oos_start, oos_end = '2013-06-30', '2024-12-31'
lookback_window = 750 
train_step = 10        # 滚动训练步长
pm_max_window = 750    
risk_window_len = 120 

# AI 训练与 CV 超参数
SEQ_LEN = 5 
HIDDEN_DIM = 16        
MAX_EPOCHS = 8         # 增加上限，交由 CV 寻找最优 epoch
BATCH_SIZE = 256       
VAL_RATIO = 0.2        # 验证集比例
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

df_factor['TradingDate'] = pd.to_datetime(df_factor['TradingDate'])
all_dates = sorted(df_factor['TradingDate'].unique())
rebalance_dates = [d for d in all_dates if pd.to_datetime(oos_start) <= d <= pd.to_datetime(oos_end)]
df_lookup = {d: group for d, group in df_factor.groupby('TradingDate')}

# --- 2. 状态存储变量 ---
weight_records = []
performance_log = [] 
oos_num_raw, oos_num_smooth, oos_denominator_list = [], [], [] 
last_w_val, last_mu_z_series, last_mu_raw_series = None, None, None
model = None

print(f"🚀 启动带时序 CV 的 LSTM+VSN-MVO 策略 (Device: {device})...")

# --- 3. 滚动预测与优化循环 ---
for i in tqdm(range(len(rebalance_dates))):
    date = rebalance_dates[i]
    date_idx = all_dates.index(date)
    
    if date_idx < lookback_window: continue
    
    # --- 3.1 动态再训练 (含验证集寻优/CV 逻辑) ---
    if i % train_step == 0 or model is None:
        train_dates = all_dates[date_idx - lookback_window : date_idx]
        full_train_df = pd.concat([df_lookup[d] for d in train_dates if d in df_lookup])[[target] + features].dropna()
        
        # 时序划分：前 80% 训练，后 20% 验证
        split_idx = int(len(full_train_df) * (1 - VAL_RATIO))
        df_t = full_train_df.iloc[:split_idx]
        df_v = full_train_df.iloc[split_idx:]
        
        def prepare_tensor(df):
            X = torch.from_numpy(df[features].values).float().unsqueeze(1).expand(-1, SEQ_LEN, -1)
            y = torch.from_numpy(df[target].values).float().view(-1, 1)
            return X, y

        X_t, y_t = prepare_tensor(df_t)
        X_v, y_v = prepare_tensor(df_v)
        X_v, y_v = X_v.to(device), y_v.to(device)

        temp_model = LSTM_VSN_Model(len(features), HIDDEN_DIM).to(device)
        optimizer = optim.Adam(temp_model.parameters(), lr=0.001)
        criterion = nn.MSELoss()
        
        train_loader = DataLoader(TensorDataset(X_t.to(device), y_t.to(device)), batch_size=BATCH_SIZE, shuffle=True)
        
        best_val_loss = float('inf')
        best_model_wts = copy.deepcopy(temp_model.state_dict())
        
        temp_model.train()
        for epoch in range(MAX_EPOCHS):
            for bx, by in train_loader:
                optimizer.zero_grad()
                criterion(temp_model(bx), by).backward()
                optimizer.step()
            
            # 验证集评估选优
            temp_model.eval()
            with torch.no_grad():
                current_val_loss = criterion(temp_model(X_v), y_v).item()
                if current_val_loss < best_val_loss:
                    best_val_loss = current_val_loss
                    best_model_wts = copy.deepcopy(temp_model.state_dict())
            temp_model.train()
        
        # 保存该阶段最优模型
        model = LSTM_VSN_Model(len(features), HIDDEN_DIM).to(device)
        model.load_state_dict(best_model_wts)
        torch.cuda.empty_cache()

    # --- 3.2 样本外预测与信号加工 ---
    current_day = df_lookup.get(date)
    if current_day is None: continue
    current_day = current_day.dropna(subset=features).copy()
    
    model.eval()
    with torch.no_grad():
        X_oos = torch.from_numpy(current_day[features].values).float().unsqueeze(1).expand(-1, SEQ_LEN, -1).to(device)
        mu_raw_output = model(X_oos).cpu().numpy().flatten()
    
    mu_z_raw_val = np.clip((mu_raw_output - np.mean(mu_raw_output)) / (np.std(mu_raw_output) + 1e-8), -3, 3)
    
    if last_mu_z_series is not None:
        mu_raw_smooth_val = 0.3 * mu_raw_output + 0.7 * last_mu_raw_series.reindex(current_day['Symbol']).fillna(0).values
        mu_z_smooth_val = 0.3 * mu_z_raw_val + 0.7 * last_mu_z_series.reindex(current_day['Symbol']).fillna(0).values
    else:
        mu_raw_smooth_val, mu_z_smooth_val = mu_raw_output, mu_z_raw_val
    
    last_mu_raw_series = pd.Series(mu_raw_smooth_val, index=current_day['Symbol'])
    last_mu_z_series = pd.Series(mu_z_smooth_val, index=current_day['Symbol'])

    # --- 3.3 MVO 资产配置 ---
    trade_candidates = current_day[current_day['is_hs300'] == True]['Symbol'].tolist()
    risk_dates = all_dates[date_idx - risk_window_len : date_idx]
    pm_dates = all_dates[max(0, date_idx - pm_max_window) : date_idx]
    
    risk_hist_df = pd.concat([df_lookup[d][['TradingDate', 'Symbol', target]] for d in risk_dates if d in df_lookup])
    pm_hist_df = pd.concat([df_lookup[d][['Symbol', target]] for d in pm_dates if d in df_lookup])
    rolling_pm_series = pm_hist_df.groupby('Symbol')[target].mean()
    
    v_matrix = risk_hist_df[risk_hist_df['Symbol'].isin(trade_candidates)].pivot(index='TradingDate', columns='Symbol', values=target).fillna(0)
    active_symbols = [s for s in trade_candidates if s in v_matrix.columns]
    
    if len(active_symbols) < 50: continue 

    mu_mvo = last_mu_z_series.loc[active_symbols].values * 0.005 
    Sigma = LedoitWolf().fit(v_matrix[active_symbols].values).covariance_
    Sigma = (Sigma + Sigma.T) / 2
    
    n = len(active_symbols)
    w = cp.Variable(n)
    curr_w = last_w_val.reindex(active_symbols).fillna(0).values if last_w_val is not None else np.ones(n)/n
    
    obj = cp.Maximize(mu_mvo @ w - 1.5 * cp.quad_form(w, Sigma) - 0.015 * cp.norm(w - curr_w, 1))
    prob = cp.Problem(obj, [cp.sum(w) == 1, w >= 0, w <= 1])
    
    try:
        prob.solve(solver=cp.OSQP, eps_abs=1e-4, eps_rel=1e-4)
        weights = np.array(w.value).flatten()
        weights[weights < 1e-4] = 0; weights /= weights.sum()
    except:
        weights = curr_w
        
    last_w_val = pd.Series(weights, index=active_symbols)
    weight_records.append(pd.DataFrame({'date': date, 'symbol': active_symbols, 'weight': weights}))
    
    # --- 3.4 统计诊断与误差计算 ---
    if target in current_day.columns:
        y_real = current_day.set_index('Symbol').reindex(active_symbols)[target].values
        daily_pm_bench = rolling_pm_series.reindex(active_symbols).fillna(0).values
        
        pred_raw = pd.Series(mu_raw_output, index=current_day['Symbol']).reindex(active_symbols).values
        ic_raw = numpy_spearmanr(pred_raw, y_real)[0, 1] if len(y_real) > 1 else 0
        
        errors_raw = y_real - pred_raw
        mse_raw = np.mean(errors_raw**2)
        mae_raw = np.mean(np.abs(errors_raw))
        rmse_raw = np.sqrt(mse_raw)
        skew_err = skew(errors_raw) if len(errors_raw) > 3 else 0
        kurt_err = kurtosis(errors_raw) if len(errors_raw) > 3 else 0
        err_sq_raw = np.sum(errors_raw**2)

        pred_smooth = last_mu_raw_series.loc[active_symbols].values
        ic_smooth = numpy_spearmanr(pred_smooth, y_real)[0, 1] if len(y_real) > 1 else 0
        err_sq_smooth = np.sum((y_real - pred_smooth)**2)
        
        bench_err_sq = np.sum((y_real - daily_pm_bench)**2)
        oos_num_raw.append(err_sq_raw)
        oos_num_smooth.append(err_sq_smooth)
        oos_denominator_list.append(bench_err_sq)

        performance_log.append({
            'Date': date, 
            'IC_Raw': ic_raw,
            'IC_Smooth': ic_smooth,
            'MSE_Raw': mse_raw,
            'MAE_Raw': mae_raw,
            'RMSE_Raw': rmse_raw,
            'Error_Skew': skew_err,
            'Error_Kurtosis': kurt_err,
            'R2_OS_Day_Raw': 1 - (err_sq_raw / (bench_err_sq + 1e-8)),
            'R2_OS_Day_Smooth': 1 - (err_sq_smooth / (bench_err_sq + 1e-8)),
            'Turnover': np.sum(np.abs(weights - curr_w)),
            'Alpha_Used': np.nan 
        })

# --- 4. 统计分析与导出 ---
if weight_records:
    total_bench = sum(oos_denominator_list) + 1e-8
    final_r2_raw = 1 - (sum(oos_num_raw) / total_bench)
    final_r2_smooth = 1 - (sum(oos_num_smooth) / total_bench)
    
    perf_df = pd.DataFrame(performance_log)
    print("\n" + "="*40)
    print(f"📊 LSTM-VSN-MVO 诊断表 (个股动态 PM 基准):")
    print(f"全样本 $R^2_{{OS}}$ (初始信号): {final_r2_raw:.6f}")
    print(f"全样本 $R^2_{{OS}}$ (平滑信号): {final_r2_smooth:.6f}")
    print(f"平均 IC (初始信号): {perf_df['IC_Raw'].mean():.4f}")
    print(f"平均 IC (平滑信号): {perf_df['IC_Smooth'].mean():.4f}")
    print(f"平均预测误差 (MSE): {perf_df['MSE_Raw'].mean():.6f}")
    print(f"平均预测误差 (MAE): {perf_df['MAE_Raw'].mean():.6f}")
    print(f"误差平均偏度: {perf_df['Error_Skew'].mean():.4f}")
    print(f"平均换手率: {perf_df['Turnover'].mean():.4f}")
    print("="*40)
    
    perf_df.to_excel('LSTM_VSN_MVO_Diagnostics.xlsx', index=False)
    pd.concat(weight_records)[lambda x: x['weight'] > 0].to_excel('LSTM_VSN_MVO_Weights.xlsx', index=False)
```
