```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

# 绘图风格美化
plt.style.use('seaborn-v0_8-muted')
plt.rcParams['font.sans-serif'] = ['Arial']
plt.rcParams['axes.unicode_minus'] = False

def plot_pure_monthly_diagnostics_with_total_r2():
    """
    针对月频纯净版回测设计的诊断绘图脚本
    手动标注全局 R2 和 IC，并计算输出【年化 ICIR】
    """
    # ---------------------------------------------------------
    # 【核心配置】：请填入模型打印的全局 Aggregate 数值
    # ---------------------------------------------------------
    GLOBAL_METRICS_MAP = {
        'Ridge': {
            'R2_OS_Raw': 0.043662,
            'IC_Raw': 0.0368
        },
        'XGBoost': {
            'R2_OS_Raw': 0.051486, 
            'IC_Raw': 0.0177
        },
        'LightGBM': {
            'R2_OS_Raw': 0.051730, 
            'IC_Raw': 0.0116
        },
        'Ensemble_CEnet': {
            'R2_OS_Raw': 0.048617, 
            'IC_Raw': 0.0368
        },
        'Ensemble_Eqw': {
            'R2_OS_Raw': 0.050191, 
            'IC_Raw': 0.0344
        }
    }

    # 1. 文件路径配置
    file_configs = {
        'Ridge': r"../weights_diagnostic/paper_use/Ridge_MVO_Monthly_Diagnostics.xlsx",
        'XGBoost': r"../weights_diagnostic/paper_use/XGB_Monthly_Diagnostics.xlsx",
        'LightGBM': r"../weights_diagnostic/paper_use/LGBM_Monthly_Diagnostics.xlsx",
        'Ensemble_CEnet': r"../weights_diagnostic/paper_use/CENet_Monthly_Diagnostics.xlsx",
        'Ensemble_Eqw': r"../weights_diagnostic/paper_use/Monthly_EW_Ensemble_Diagnostics.xlsx"
    }

    # 2. 核心指标配置
    metrics = {
        'IC_Raw': {
            'label': 'Spearman IC',
            'title': 'Monthly Information Coefficient (IC) & Annualized ICIR',
            'baseline': 0,
            'color': '#1f77b4'
        },
        'R2_OS_Raw': {
            'label': 'Out-of-Sample $R^2$',
            'title': 'Monthly $R^2_{OS}$ Performance',
            'baseline': 0.005,
            'color': '#2ca02c'
        }
    }

    WINDOW = 12 

    for model_name, file_path in file_configs.items():
        if not os.path.exists(file_path):
            print(f"⏩ 跳过 {model_name}: 文件未找到")
            continue

        df = pd.read_excel(file_path)
        df['Date'] = pd.to_datetime(df['Date'])
        df.sort_values('Date', inplace=True)

        for col, info in metrics.items():
            if col not in df.columns:
                continue

            plt.figure(figsize=(12, 6))
            
            # --- 逻辑计算 ---
            rolling_avg = df[col].rolling(WINDOW, min_periods=1).mean()
            
            # 确定全局 Aggregate 数值
            if model_name in GLOBAL_METRICS_MAP and col in GLOBAL_METRICS_MAP[model_name]:
                total_sample_val = GLOBAL_METRICS_MAP[model_name][col]
            else:
                total_sample_val = df[col].mean()

            # --- 【核心修改：年化 ICIR 计算逻辑】 ---
            ann_icir_text = ""
            if col == 'IC_Raw':
                ic_mean = df[col].mean()
                ic_std = df[col].std()
                # 年化系数为 sqrt(12)
                ann_icir = (ic_mean / ic_std) * np.sqrt(12) if ic_std != 0 else 0
                ann_icir_text = f" | Ann. ICIR: {ann_icir:.4f}"
            
            # --- 绘图步骤 ---
            # 1. 绘制月度原始波动
            plt.plot(df['Date'], df[col], color='lightgray', lw=0.8, alpha=0.4, label='Monthly Raw Value')
            
            # 2. 绘制滚动趋势线
            plt.plot(df['Date'], rolling_avg, color=info['color'], lw=2.5, label=f'{WINDOW}M Rolling Mean')
            
            # 3. 填充滚动标准差区间
            rolling_std = df[col].rolling(WINDOW, min_periods=1).std()
            plt.fill_between(df['Date'], rolling_avg - rolling_std, rolling_avg + rolling_std, 
                             color=info['color'], alpha=0.1)

            # 4. 绘制全局水平线（包含年化 ICIR 标注）
            line_label = f'Global Aggregate: {total_sample_val:.6f}{ann_icir_text}'
            plt.axhline(total_sample_val, color='red', linestyle='--', lw=2, label=line_label)
            
            # 5. 绘制学术基准线
            plt.axhline(0, color='black', lw=1.2, alpha=0.8)
            if col == 'R2_OS_Raw':
                plt.axhline(0.005, color='darkorange', linestyle=':', lw=1.5, 
                            label='Campbell (2008) Sig. Threshold (0.5%)')

            # --- 图表美化 ---
            plt.title(f'{model_name} - {info["title"]}', fontsize=16, fontweight='bold', pad=20)
            plt.ylabel(info['label'], fontsize=12)
            plt.xlabel('Rebalancing Date', fontsize=12)
            
            plt.legend(loc='upper left', frameon=True, fontsize=10, ncol=2)
            plt.grid(True, linestyle=':', alpha=0.6)
            
            # 智能缩放
            y_data = rolling_avg.dropna()
            if not y_data.empty:
                y_min = min(total_sample_val, df[col].min(), y_data.min())
                y_max = max(total_sample_val, df[col].max(), y_data.max())
                padding = (y_max - y_min) * 0.2 if y_max != y_min else 0.05
                plt.ylim(y_min - padding, y_max + padding)

            plt.tight_layout()
            
            # 保存
            save_name = f'Diagnostic_{model_name}_{col}.png'
            plt.savefig(save_name, dpi=300)
            plt.close()
            
            output_msg = f"✅ 已保存 {model_name} 的 {col} 诊断图。"
            if ann_icir_text:
                output_msg += f" 计算得到 {ann_icir_text}"
            print(output_msg)

if __name__ == "__main__":
    plot_pure_monthly_diagnostics_with_total_r2()
    print("\n✨ 诊断图更新完成。IC 图表中已包含【年化 ICIR】标注。")
```
