```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 设置绘图风格
plt.style.use('seaborn-v0_8')
plt.rcParams['font.sans-serif'] = ['Arial'] 
plt.rcParams['axes.unicode_minus'] = False

def plot_all_diagnostics():
    # --- [预留空间：请在此填写你的 Excel 统计结果] ---
    # 填入 Excel 诊断表最后的平均值
    target_stats_comparison = {
        'Ridge': {
            'R2_Raw': 0.001079, 'R2_Smooth': 0.001228, 
            'IC_Raw': 0.0233, 'IC_Smooth': 0.0192
        },
        'XGBoost': {
            'R2_Raw': -0.001461, 'R2_Smooth': 0.000408, 
            'IC_Raw': 0.0187, 'IC_Smooth': 0.0175
        },
        'LightGBM': {
            'R2_Raw': -0.002985, 'R2_Smooth': -0.000232, 
            'IC_Raw': 0.0181, 'IC_Smooth': 0.0174
        },
        'Ensemble_Eqw': {
            'R2_Raw': 0.001593, 'R2_Smooth': 0.001517, 
            'IC_Raw': 0.0244, 'IC_Smooth': 0.0211
        },
        'Ensemble_CEnet': {
            'R2_Raw': 0.001679, 'R2_Smooth': 0.001563, 
            'IC_Raw': 0.0254, 'IC_Smooth': 0.0206
        },
        'LSTM_VSN': {
            'R2_Raw': -0.001838, 'R2_Smooth': 0.000401, 
            'IC_Raw': 0.0180, 'IC_Smooth': 0.0145
        }
    }

    # 1. 文件配置 (保持路径不变)
    file_configs = {
        'Ridge': r"../weights_diagnostic/old/岭回归3/Ridge_MVO_DynamicPM_Diagnostics.xlsx",
        'XGBoost': r"../weights_diagnostic/old/XGBoost3/XGB_DynamicPM_Diagnostics.xlsx",
        'LightGBM': r"../weights_diagnostic/old/LightGBM3/LGBM_DynamicPM_Diagnostics.xlsx",
        'Ensemble_CEnet': r"../weights_diagnostic/old/集成模型3/CENet_Dynamic_Diagnostics.xlsx",
        'Ensemble_Eqw': r"../weights_diagnostic/old/集成模型3/EW_Ensemble_Diagnostics.xlsx",
        'LSTM_VSN': r"../weights_diagnostic/old/LSTM_VSN/LSTM_VSN_MVO_Diagnostics.xlsx"
    }

    # 2. 指标配置 (针对你的 Excel 列名进行精准对齐)
    metrics = {
        'IC': {
            'label': 'Information Coefficient (IC)',
            'raw_col': 'IC_Raw',
            'smooth_col': 'IC_Smooth',
            'title_suffix': 'IC Stability Analysis (Raw vs. Smooth)'
        },
        'R2': {
            'label': 'Out-of-Sample $R^2$',
            'raw_col': 'R2_OS_Day_Raw',
            'smooth_col': 'R2_OS_Day_Smooth',
            'title_suffix': 'OOS $R^2$ Stability Analysis (Raw vs. Smooth)'
        }
    }

    for model_name, file_path in file_configs.items():
        try:
            df = pd.read_excel(file_path)
            df['Date'] = pd.to_datetime(df['Date'])
            df.sort_values('Date', inplace=True)
        except Exception as e:
            print(f"⚠️ Skip {model_name}: {e}")
            continue

        for m_key, m_info in metrics.items():
            raw_col = m_info['raw_col']
            smooth_col = m_info['smooth_col']
            
            if raw_col not in df.columns or smooth_col not in df.columns:
                print(f"❌ Missing columns for {model_name} {m_key}")
                continue

            # --- 绘图逻辑 ---
            plt.figure(figsize=(12, 7))
            
            # 计算 20日滚动均值以平滑绘图趋势
            raw_rolling = df[raw_col].rolling(20).mean()
            smooth_rolling = df[smooth_col].rolling(20).mean()
            
            # 绘制 Raw 曲线 (浅灰色背景线)
            plt.plot(df['Date'], raw_rolling, color='gray', lw=1.2, alpha=0.35, 
                     label=f'20D Rolling {m_key} (Raw)')
            
            # 绘制 Smooth 曲线 (主深蓝线)
            plt.plot(df['Date'], smooth_rolling, color='tab:blue', lw=2.2, 
                     label=f'20D Rolling {m_key} (Smoothed)')
            
            # 填充平滑信号阴影
            plt.fill_between(df['Date'], smooth_rolling, 0, color='tab:blue', alpha=0.08)
            
            # --- 均值线逻辑：优先使用 target_stats_comparison 中的精确值 ---
            if model_name in target_stats_comparison:
                m_data = target_stats_comparison[model_name]
                display_mean_raw = m_data[f'{m_key}_Raw']
                display_mean_smooth = m_data[f'{m_key}_Smooth']
            else:
                display_mean_raw = df[raw_col].mean()
                display_mean_smooth = df[smooth_col].mean()
            
            # 原始均值线 (橙色虚线)
            plt.axhline(display_mean_raw, color='darkorange', linestyle=':', lw=1.5, 
                        label=f'Mean {m_key} (Raw): {display_mean_raw:.6f}')
            
            # 平滑均值线 (红色长虚线)
            plt.axhline(display_mean_smooth, color='red', linestyle='--', lw=1.8, 
                        label=f'Mean {m_key} (Smooth): {display_mean_smooth:.6f}')
            
            # 零轴
            plt.axhline(0, color='black', lw=1, alpha=0.5)
            
            # 格式化
            plt.title(f'{model_name} - {m_info["title_suffix"]}', fontsize=15, pad=15)
            plt.ylabel(m_info['label'], fontsize=12)
            plt.xlabel('Date', fontsize=12)
            plt.legend(loc='upper left', frameon=True, ncol=2, fontsize=10)
            plt.grid(True, linestyle=':', alpha=0.6)
            
            # 坐标轴智能缩放逻辑
            all_vals = pd.concat([raw_rolling, smooth_rolling]).dropna()
            if not all_vals.empty:
                y_min, y_max = all_vals.min(), all_vals.max()
                y_min = min(y_min, display_mean_raw, display_mean_smooth, 0)
                y_max = max(y_max, display_mean_raw, display_mean_smooth)
                
                margin = (y_max - y_min) * 0.25 if (y_max - y_min) > 0 else 0.05
                plt.ylim(y_min - margin, y_max + margin)

            plt.tight_layout()
            
            # 保存 (维持原有名字格式)
            save_name = f'Diagnostic_{model_name}_{m_key}_Smooth.png'
            plt.savefig(save_name, dpi=300)
            plt.close()
            print(f"✅ Saved: {save_name} | Smooth Mean: {display_mean_smooth:.6f}")

# 执行
if __name__ == "__main__":
    plot_all_diagnostics()
    print("\n✨ Done! All comparison plots generated using original filenames.")
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import ast

# --- 绘图风格配置 ---
plt.style.use('seaborn-v0_8')
plt.rcParams['font.sans-serif'] = ['Arial']
plt.rcParams['axes.unicode_minus'] = False

def plot_factor_evolution_individual(file_path, model_name, col_index):
    try:
        # 1. 读取数据
        df = pd.read_excel(file_path)
        
        # 2. 解析日期和指定的 Top_5_Features 列
        # iloc[:, col_index]：Ridge传5(第六列)，其他传4(第五列)
        date_col = pd.to_datetime(df['Date'])
        feature_col = df.iloc[:, col_index] 
        
        feature_records = []
        for i, row in enumerate(feature_col):
            try:
                # 处理字符串格式的列表，如 "['f1', 'f2']"
                features = ast.literal_eval(str(row))
                # 评分机制：排名第1给5分，第5给1分
                for rank, f in enumerate(features):
                    feature_records.append({
                        'Year': date_col.iloc[i].year,
                        'Feature': f,
                        'Score': 5 - rank
                    })
            except:
                continue
        
        if not feature_records:
            print(f"⚠️ {model_name} 日志中未提取到有效数据，请检查第 {col_index+1} 列内容。")
            return

        f_df = pd.DataFrame(feature_records)
        
        # 3. 按“因子-年份”汇总得分
        pivot_df = f_df.groupby(['Feature', 'Year'])['Score'].sum().unstack(fill_value=0)
        
        # 4. 筛选该模型总分最高的前 15 个因子进行展示
        top_15 = pivot_df.sum(axis=1).sort_values(ascending=False).head(15).index
        plot_data = pivot_df.loc[top_15]

        # 5. 绘图
        plt.figure(figsize=(12, 8))
        sns.heatmap(plot_data, annot=True, cmap='YlGnBu', fmt='.0f', linewidths=0.5)
        
        plt.title(f'{model_name} - Top 15 Factors Temporal Evolution', fontsize=15)
        plt.ylabel('Alpha Factors')
        plt.xlabel('Year')
        
        plt.tight_layout()
        save_name = f'Factor_Evolution_{model_name}.png'
        plt.savefig(save_name, dpi=300)
        plt.show()
        print(f"✅ {model_name} 因子演化图已生成: {save_name}")

    except Exception as e:
        print(f"❌ 处理 {model_name} 时出错: {e}")

# --- 配置字典：名称, 路径, 列索引(从0开始算) ---
configs = [
    {
        'name': 'Ridge',
        'path': r"../weights_diagnostic/old/岭回归/Backtest_Ridge_Diagnostics.xlsx",
        'col': 5  # 第六列
    },
    {
        'name': 'XGBoost',
        'path': r"../weights_diagnostic/old/XGBoost/Backtest_XGB_Diagnostic_Log.xlsx",
        'col': 4  # 第五列
    },
    {
        'name': 'LightGBM',
        'path': r"../weights_diagnostic/old/LightGBM/Backtest_LGBM_Diagnostic_Log.xlsx",
        'col': 4  # 第五列
    }
]

# --- 循环执行 ---
for cfg in configs:
    plot_factor_evolution_individual(cfg['path'], cfg['name'], cfg['col'])
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import ast
from collections import Counter

# --- 绘图风格配置 ---
plt.style.use('seaborn-v0_8')
plt.rcParams['font.sans-serif'] = ['Arial']
plt.rcParams['axes.unicode_minus'] = False

def plot_triple_factor_comparison(configs):
    model_data = {}
    
    for cfg in configs:
        model_name = cfg['name']
        path = cfg['path']
        col_index = cfg['col']
        
        try:
            df = pd.read_excel(path)
            # 根据索引提取对应的 Top_5_Features 列
            feature_series = df.iloc[:, col_index].dropna()
            
            all_features = []
            for row in feature_series:
                try:
                    # 解析字符串格式的列表
                    features = ast.literal_eval(str(row))
                    all_features.extend(features)
                except:
                    continue
            
            # 统计频率：计算因子在回测期内进入 Top 5 的总次数
            counts = Counter(all_features)
            # 归一化：出现次数 / 总样本数 (代表该因子被该模型看重的时间比例)
            total_epochs = len(df)
            model_data[model_name] = {k: v / total_epochs for k, v in counts.items()}
            print(f"✅ {model_name} 数据解析完成")
            
        except Exception as e:
            print(f"⚠️ 读取 {model_name} 失败: {e}")

    # 1. 整合三个模型的数据到同一个 DataFrame
    combined_df = pd.DataFrame(model_data).fillna(0)
    
    # 2. 排序与筛选：取三个模型综合频率最高的前 20 个因子
    top_factors = combined_df.mean(axis=1).sort_values(ascending=False).head(20).index
    plot_df = combined_df.loc[top_factors]

    # 3. 绘制热力图 (三列对比)
    plt.figure(figsize=(10, 12))
    # 使用 YlGnBu 渐变色，数值越大颜色越深
    sns.heatmap(plot_df, annot=True, cmap='YlGnBu', fmt='.2f', linewidths=0.5)
    
    plt.title('Factor Selection Preference: Ridge vs XGBoost vs LightGBM', fontsize=15)
    plt.ylabel('Alpha Factors (Top 20 by Frequency)', fontsize=12)
    plt.xlabel('Models', fontsize=12)
    
    plt.tight_layout()
    save_name = 'Triple_Factor_Comparison_Heatmap.png'
    plt.savefig(save_name, dpi=300)
    plt.show()
    print(f"📊 最终对比热力图已保存为: {save_name}")

# --- 配置文件路径与列索引 ---
configs = [
    {
        'name': 'Ridge',
        'path': r"../weights_diagnostic/old/岭回归/Backtest_Ridge_Diagnostics.xlsx",
        'col': 5  # 第六列
    },
    {
        'name': 'XGBoost',
        'path': r"../weights_diagnostic/old/XGBoost/Backtest_XGB_Diagnostic_Log.xlsx",
        'col': 4  # 第五列
    },
    {
        'name': 'LightGBM',
        'path': r"../weights_diagnostic/old/LightGBM/Backtest_LGBM_Diagnostic_Log.xlsx",
        'col': 4  # 第五列
    }
]

# --- 执行 ---
if __name__ == "__main__":
    plot_triple_factor_comparison(configs)
```
